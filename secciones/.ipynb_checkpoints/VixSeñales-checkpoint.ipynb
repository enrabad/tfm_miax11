```python
# Se optimizan los precios con un delta operador para conseguir unos drawdowns objetivo
# y se generan los trades con dichos precios óptimos con el siguiente valor no visto siguiendo 
# con un walk forward de ventana expansiva con un único dato de test (el valor siguiente)

# El problema de esta versión es que la optimización es direccional, y depende de si el periodo es alcista o bajista
# La parte positiva es que las compras tienen menos volatilidad, con lo que es probable que cuando el mercado baje, 
# invertir en compras mejore al mercado

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
from itertools import product
import random
import time
from pathlib import Path

def cargar_datos (desde, hasta):
    '''
    
    '''

    spy_datos = yf.download("SPY", start=desde, end=hasta)
    vix_datos = yf.download("^VIX", start=desde, end=hasta)

    datos_all = pd.concat([spy_datos.Close, vix_datos.Close], axis=1).dropna()
    datos_all.columns = ["SPY","VIX"]
    datos_all["rets"] = datos_all.SPY.pct_change()

    # Indicadores derivados del VIX
    datos_all["FVIX"] = datos_all.VIX
    datos_all["FRET"] = datos_all.rets
    datos_all["FVOLVIX"] = np.abs(datos_all.rets) - datos_all.VIX/(np.sqrt(252)*100)
    datos_all["FSENVIX"] = datos_all.rets - datos_all.VIX/(np.sqrt(252)*100)
    
    datos_all.dropna(inplace=True)

    return datos_all

desde = "2000-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")
hasta = "2024-11-20"
datos = cargar_datos (desde, hasta)
```

```python
def msignal (datos, activo, ind, periodo, nvar, op):

    if periodo == 0:
        if op == ">":
            señal = datos[ind] >= nvar*datos[ind].std()
        else:
            señal = datos[ind] < nvar*datos[ind].std()
    else:
        if op == ">":
            señal = datos[ind] >= (datos[ind].rolling(periodo).mean() + nvar*datos[ind].rolling(periodo).std())
        else:
            señal = datos[ind] < (datos[ind].rolling(periodo).mean() - nvar*datos[ind].rolling(periodo).std())
    
    return (señal)

def vsignal (datos, activo, ind, periodo, nvar, op):

    if periodo == 0:
        if op == ">":
            señal = datos[ind] >= np.abs(datos[activo].pct_change()) * nvar
        else:
            señal = datos[ind] < np.abs(datos[activo].pct_change()) * nvar
    else:
        if op == ">":
            señal = datos[ind] >= (datos[ind].rolling(periodo).std()) * nvar
        else:
            señal = datos[ind] < (datos[ind].rolling(periodo).std()) * nvar
    
    return (señal)
```

```python

def wfo (datos, inidatos, ventana_train, ventana_test, indicador, activo, params, signo=1, fseñal=msignal, expansiva=False):

    df_result = pd.DataFrame()
    if inidatos + ventana_train > len(datos):
        return

    datos["rets"] = np.log(datos[activo]).diff()
    for i in range(inidatos + ventana_train, len(datos) - ventana_test, ventana_test):
        if expansiva:
            datos_train = datos.iloc[:i]
        else:
            datos_train = datos.iloc[i - ventana_train:i]

        datos_test = datos.iloc[i:i + ventana_test]
        maxrent = -999
        maxidx = -1
        for idx, param in enumerate(params):
            datos_train["señal"] = fseñal(datos_train, activo, indicador, param[0], param[1], param[2])
            rentmed = (datos_train.señal * signo * datos_train["rets"].shift(-1)).mean()
            if rentmed > maxrent:
                maxrent = rentmed
                maxidx = idx

        param = params[maxidx]
        datos_test["señal"] = msignal(datos_test, activo, indicador, param[0], param[1], param[2])
        df_result = pd.concat([df_result, datos_test])
            
    return df_result, param

params1 = [0, 10, 25, 50, 100, 200]
params2 = [0, 0.5, 1, 1.5, 2, 2.5]
params3 = ["<", ">"]
params = list(product(params1, params2, params3))
result, param = wfo (datos, inidatos=200, ventana_train=250*7, ventana_test=250,
                     indicador="FSENVIX", activo="SPY", params=params, signo=1)
(1 + result.señal*result.SPY.pct_change().shift(-1)).cumprod().plot()
(1 + result.SPY.pct_change().shift(-1)).cumprod().plot()
```

```python

ini = time.time()

def designia_qt (datos_qt, datos, señales, periodos, result, numqt=10):
    
    if datos.index[-1] < datos_quantil.index[-1]:
        return
    
    qdatos = []
    for j in range(len(señales)):
        qdatos.append(np.array(datos_quantil[señales[j]].quantile([i / 10 for i in range(1, numqt+1)])))

    df_qdatos = pd.DataFrame(np.array(qdatos).T)
    df_qdatos.columns = señales
    
    qdatosY = np.array(datos_quantil[result].quantile([i / 10 for i in range(1, numqt+1)]))

    fechas = []
    max_periodo = periodos[-1]
    #dias, peridos, deciles, señal=[vixsi, retvix, vix]
    ldatos = []
    ldatosY = []

    for i in range(len(datos)-max_periodo):
        mdatos = np.zeros((numqt, numqt, numsig))
        mdatosY = np.zeros((numqt))
        for j in range(len(periodos)):
            datos_frec = datos.iloc[i+max_periodo-periodos[j]:i+max_periodo]
            for k in range(len(qdatos)):
                if k == 0:
                    for s in range(numsig):
                        mdatos[j, k, s] = len(datos_frec.loc[(datos_frec[df_qdatos.columns[s]] <= df_qdatos.iloc[k,s])])
                    if periodos[j]==1:
                        mdatosY[k] = len(datos_frec.loc[(datos_frec[result] < qdatosY[k])])

                else:
                    for s in range(numsig):
                        mdatos[j, k, s] = len(datos_frec.loc[(datos_frec[df_qdatos.columns[s]] > df_qdatos.iloc[k-1,s]) &
                                                      (datos_frec[df_qdatos.columns[s]] <= df_qdatos.iloc[k,s])])
                    if periodos[j]==1:
                        mdatosY[k] = len(datos_frec.loc[(datos_frec[result] <= qdatosY[k]) &
                                                          (datos_frec[result] > qdatosY[k-1])])
                if k == len(qdatos)-1:
                    for s in range(numsig):
                        mdatos[j, k, s] = (mdatos[j, k, s] +
                          len(datos_frec.loc[(datos_frec[df_qdatos.columns[s]] > df_qdatos.iloc[k,s])]))
                    if periodos[j]==1:
                        mdatosY[k] = mdatosY[k] + len(datos_frec.loc[(datos_frec[result] > qdatosY[k])])

            for s in range(len(df_qdatos.columns)):
                mdatos[j,:,s] = mdatos[j,:,s]/mdatos[j,:,s].sum()
            if periodos[j]==0:
                mdatosY = mdatosY/mdatosY.sum()

        ldatos.append(mdatos)
        ldatosY.append(mdatosY)
        fechas.append(datos.iloc[i+max_periodo-periodos[j]:i+max_periodo].index[-1])
            
    return np.array(ldatos), np.array(ldatosY), fechas

print ("tiempo creación datos train_test:", time.time()-ini)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, MaxPooling2D

# Dimensiones de entrada y salida
input_shape = (10, 10, canales)  # Matriz de píxeles 10x10 con 3 canales (color)
num_classes = 10           # 10 clases de salida

# Crear el modelo de la red convolucional
model = Sequential([
    # Capa convolucional con 32 filtros, kernel de 3x3 y activación ReLU
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    # Capa de pooling para reducir dimensiones
    MaxPooling2D(pool_size=(2, 2)),
    # Aplanar las características
    Flatten(),
    # Capa totalmente conectada con 64 neuronas y activación ReLU
    Dense(64, activation='relu'),
    # Capa de salida con activación softmax para clasificación en 10 clases
    Dense(num_classes, activation='softmax')
])

# Compilar el modelo
model.compile(
    optimizer='adam',                # Optimizador Adam
    loss='sparse_categorical_crossentropy',  # Pérdida para clasificación multiclase
    metrics=['accuracy']             # Métrica de precisión
)

# Resumen del modelo
model.summary()
```

```python

def optimizar (trades, opgain):
    
    qtrades = trades.copy()
    qtrades["vpoutl"] = 0
    if opgain=="vpgain":
        #trades.loc[qtrades[opgain] < qtrades[opgain].quantile(0.05),"vpoutl"] = 1
        qtrades.loc[qtrades["retsig"] < qtrades["retsig"].quantile(0.10),"vpoutl"] = 1
        minmax = qtrades.loc[qtrades.vpoutl==1].qvixsi.describe(percentiles=[0.25, 0.75])
        optmin = minmax[4]
        optmax = minmax[6]
    if opgain=="cpgain":
        #trades.loc[qtrades[opgain] < qtrades[opgain].quantile(0.05),"vpoutl"] = 1
        qtrades.loc[qtrades["retsig"] < qtrades["retsig"].quantile(0.10),"vpoutl"] = 1
        minmax = qtrades.loc[qtrades.vpoutl==0].qvixsi.describe()
        optmin = minmax[3]
        optmax = minmax[7]
    if opgain=="vcgain":
        qtrades.loc[qtrades["retsig"] > qtrades["retsig"].quantile(0.90),"vpoutl"] = 1
        minmax = qtrades.loc[qtrades.vpoutl==1].qvixsi.describe(percentiles=[0.25, 0.75])
        optmin = minmax[4]
        optmax = minmax[6]
    if opgain=="ccgain":
        qtrades.loc[qtrades["retsig"] > qtrades["retsig"].quantile(0.90),"vpoutl"] = 1
        minmax = qtrades.loc[qtrades.vpoutl==0].qvixsi.describe()
        optmin = minmax[3]
        optmax = minmax[7]

    return optmax, optmin

```

```python
# Optimización utilizando datos de train (el 80%) y datos de test (el 20% restante)

def optimizar_train_test (datos, df_result, opgain, qvixsi, ini_test, fin_test, in_test_train=False):
    '''
    
    '''

    if in_test_train:
        datos_train = datos.iloc[0:fin_test].copy()
    else:
        datos_train = datos.iloc[0:ini_test].copy()
        
    datos_test = datos.iloc[ini_test:fin_test].copy()

    if qvixsi != "None":      
        datos_train["qvixsi"] = datos_train[qvixsi]
        datos_test["qvixsi"] = datos_test[qvixsi]

        optmax, optmin = optimizar (datos_train, opgain=opgain)

        #if len(df_result) > 0:
        #    df_result = df_result[df_result.SIGNAL==qvixsi]
        #    num_regs = len(df_result)
        #    if num_regs > 0:
        #        optmax = (df_result.MAX.sum() + optmax)/(num_regs + 1)
        #        optmin = (df_result.MIN.sum() + optmin)/(num_regs + 1)

        qvtrades = datos_test
        if qvixsi =="qretvix" and opgain =="cpgain":
            qvtrades.loc[qvtrades.qvixsi <= 0] = 0
        else:
            qvtrades.loc[(qvtrades.qvixsi <= optmax) &
             (qvtrades.qvixsi >= optmin), opgain] = 0
    else:
        qvtrades = datos_test
        optmax, optmin = 1, 0 

    df_result = pd.DataFrame()
    df_result = pd.DataFrame([[qvixsi, opgain, ini_test, fin_test, optmax, optmin]])
    df_result.columns = ["SIGNAL", "OPGAIN", "INI_TEST", "FIN_TEST", "MAX","MIN"]

    return df_result, qvtrades
```

```python
#Para un test del 10% en cualquier lugar del periodo hacia adelante desde el 40% de los datos

def optimizar_wfe (trades, opgain, min_dias_train=250, dias_test=20):
    
    df_result = pd.DataFrame()
    df_None = pd.DataFrame()
    df_qrvix = pd.DataFrame()
    df_qvix = pd.DataFrame()
    df_qret = pd.DataFrame()
    df_qretvix = pd.DataFrame()

    maxregs = len (trades)
    ini_test = min_dias_train
    while (ini_test < maxregs):
        fin_test = ini_test + dias_test
        if fin_test >= maxregs:
            fin_test = maxregs


        result, strades = optimizar_train_test (trades, df_result, opgain, qvixsi="qvix",
                                                ini_test=ini_test, fin_test=fin_test)
        df_result  = pd.concat([df_result, result])
        df_qvix = pd.concat([df_qvix,strades])

        result, strades = optimizar_train_test (trades, df_result, opgain, qvixsi="qrvix",
                                                ini_test=ini_test, fin_test=fin_test)
        df_result  = pd.concat([df_result, result])
        df_qrvix = pd.concat([df_qrvix,strades])
        

        result, strades = optimizar_train_test (trades, df_result, opgain, qvixsi="qretvix",
                                                ini_test=ini_test, fin_test=fin_test)
        df_result  = pd.concat([df_result, result])
        df_qretvix = pd.concat([df_qretvix,strades])
        
        ini_test = fin_test

    return df_result, {"vix":df_qvix, "rvix":df_qrvix, "retvix":df_qretvix}
```
