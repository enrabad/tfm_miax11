## Analizar alpha generado por la señal VIXSI en el activo SPY

Una forma de validar la señal es ver cuanto alpha genera la estrategia que utiliza la señal para invertir diariamente en largo (vixsi=1) o en corto (vixsi=-1) en un activo respecto a mantenerse invertido en el propio activo.

Incialmente se utilizará como activo el ETF que replica al SP500 (SPY).

la fórmula del alpha se basará en el alpha de Jensen (asignando 0 a la rentabilidad libre de riesgo):

alpha = rentabilidad señal - Beta_señal_activo * rentabilidad activo, y se calculará en periodos anuales.

### alpha_vixsi_spy

Una forma de validar la señal es ver cuanto alpha genera la estrategia que utiliza la señal para invertir en largo (vixsi=1) o en corto (vixsi=-1) un activo respecto a mantenerse invertido en el activo.

la fórmula del alpha se basará en el alpha de Jensen (asignando 0 a la rentabilidad libre de riesgo):

rentabilidad señal - Beta_señal_activo * rentabilidad activo

El alpha se calculará por año, y se evaluará la media y el ratio de sharpe del alpha anual.

Los datos del SPY se recogerán de yfinance y se ajustarán a los dividendos.


```python
def f_alpha(datos, estrategia="estrategia", benchmark="SPY"):
    '''

    '''

    srets = np.log(datos[estrategia]).diff().fillna(0)
    brets = np.log(datos[benchmark]).diff().fillna(0)

    #w_brets = brets.loc[srets.index]
    cov_rets = np.cov(srets.T, brets)[0,1:]
    var_brets = brets.var()
    beta = cov_rets / var_brets
    rent_srets = srets.sum()
    rent_brets = brets.sum()
    alpha = rent_srets - beta * rent_brets    
    return (alpha)

benchmark = "SPY"

desde = "2008-01-01"
hasta = "2024-01-01"

#Los datos históricos del VIXSI se extraen de un fichero
df_vixsi = pd.read_csv("../datos/VIXSIyf.csv", index_col=0)
df_vixsi.index =  pd.to_datetime(df_vixsi.index)

#Los datos del SPY se recogerán de yfinance y se ajustarán a los dividendos
#spy_datos = yf.download("SPY", start=desde, end=hasta, auto_adjust=True)
spy_datos = yf.download("KLAC", start=desde, end=hasta, auto_adjust=True)

datos = pd.concat([spy_datos.Close, df_vixsi.VIXSI], axis=1).dropna()
datos.columns = ["SPY","vixsi"]

#Se calculan los retornos acumulados de las estrategias holding y vixsi
datos["estrHolding"] = (1 + datos[benchmark].pct_change()).cumprod()
datos["estrvixsi"] = (1 + datos.vixsi.shift(1) * datos[benchmark].pct_change()).cumprod()

#Se generan los alphas por año
años = sorted(datos.index.year.unique())
alpha = np.zeros(len(años))
for i in range(len(años)):
    alpha[i] = f_alpha(datos.loc[str(años[i])],estrategia="estrvixsi")
    
alpha = pd.DataFrame(alpha, index=años)

print("media anual del alpha:", alpha.mean().values)

plt.figure(1)
alpha.plot()
plt.title('Alpha de la estrategia VIXSI respecto al SPY')
plt.axhline(y=0, color='b', linestyle='--')
plt.show()

plt.figure(2)
estrategias = datos.filter(regex='^estr')
estrategias.plot()
plt.title('Equity estrategias')
plt.show()
```

```python

def calcula_estrategia_par (datos, estrategia, p_dwd1, p_dwd2):
    max252 = datos.SPY.rolling(252).max()
    datos["dwd"] = (max252 - datos.SPY)/max252*100
    datos["sigdwd1"] = (datos["dwd"] < p_dwd1) * 1
    datos["sigdwd2"] = (datos["dwd"] > p_dwd2) * 1
    datos[estrategia] = (1 + (datos.SPY.pct_change() * datos.sigdwd1.shift(1) + 
        (1 - datos.sigdwd1.shift(1))*(1 - datos.sigdwd2.shift(1))*datos.vixsi.shift(1) * datos.SPY.pct_change() +
        (1 - datos.sigdwd1.shift(1))*datos.sigdwd2.shift(1)*((datos.vixsi.shift(1) - 1)/2) * datos.SPY.pct_change())).cumprod()

def optimiza_par_estrategia (datos, estrategia):
    max_sharpe = 0
    l_dwd1 = [-1,2,3,5,7]
    l_dwd2 = [10,20,30,40,50,100]
    for p_dwd1 in l_dwd1:
        for p_dwd2 in l_dwd2:
            calcula_estrategia_par (datos, estrategia, p_dwd1, p_dwd2)
            _rets = np.log(datos[estrategia]).diff()
            _std = _rets.std()
            rsharpe = _rets.mean()/_std
            if rsharpe > max_sharpe:
                pm_dwd1 = p_dwd1
                pm_dwd2 = p_dwd2
                max_sharpe = rsharpe
    return pm_dwd1, pm_dwd2

estrategia = "estrvixsi_par"
pm_dwd1, pm_dwd2 = optimiza_par_estrategia(datos, estrategia)
print ("parámetros óptimos:", pm_dwd1, pm_dwd2)
calcula_estrategia_par (datos, estrategia, pm_dwd1, pm_dwd2)
estrategias = datos.filter(regex='^estr')
estrategias.plot()
plt.title('Equity estrategias')
plt.show()
```
