```python
# El dataset de entrada son los deciles valores normalizados de 1 día a 10 días el dataset de salida son los deciles de retsig
# Se optimiza el modelo CNN

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time
from pathlib import Path
import pickle

from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense, LSTM, Flatten, BatchNormalization, MaxPooling1D, Attention
from tensorflow.keras.models import Model

def cargar_datos (desde, hasta):
    '''
    
    '''

    spy_datos = yf.download("SPY", start=desde, end=hasta)
    vix_datos = yf.download("^VIX", start=desde, end=hasta)

    datos_all = pd.concat([spy_datos.Close, vix_datos.Close], axis=1).dropna()
    datos_all.columns = ["SPY","VIX"]
    datos_all["rets"] = datos_all.SPY.pct_change()

    # Indicadores para deltavix
    datos_all["FVIX"] = datos_all.VIX
    datos_all["FRETVIX"] = datos_all.VIX.pct_change()
    datos_all["FRET"] = datos_all.rets
    datos_all["FRVIX"] = np.abs(datos_all.rets) - datos_all.VIX/(np.sqrt(256)*100)
    datos_all["FVIXSI"] = datos_all.rets - datos_all.VIX/(np.sqrt(256)*100)
    #datos_all["vixsi"] = np.where(datos_all.rets - datos_all.VIX/(np.sqrt(256)*100) > 0, -1, 1)
    datos_all["vixsi"] = np.where(datos_all.rets - datos_all.VIX/(np.sqrt(256)*100) > 0, 0, 1)
    datos_all["qretvix"] = datos_all.FRETVIX
    datos_all["qret"] = datos_all.FRET
    datos_all["qrvix"] = datos_all.FRVIX
    datos_all["qvix"] = datos_all.FVIX
    datos_all["retsig"] = datos_all.rets.shift(-1)

    datos_all.dropna(inplace=True)

    return datos_all

```

```python
desde = "2008-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")
hasta = "2024-11-20"
datos = cargar_datos (desde, hasta)
```

```python
#Conversión a 0-1 según rank
def datos_train_test (datos, desde_train, hasta_train, desde_test, hasta_test):
    
    datos_train = datos.loc[desde_train:hasta_train]
    datos_train["QVIXSI"] = datos.FVIXSI.rank(pct=True)
    datos_train["QVIX"] = datos.FVIX.rank(pct=True)
    
    datos_test = datos.loc[desde_test:hasta_test]
    datos_test["QVIXSI"] = None
    datos_test["QVIX"] = None

    for i in datos_test.index:
        vixsi = datos_test.loc[i].FVIXSI
        if vixsi >= datos_train.FVIXSI.min():
            idq = datos_train[datos_train.FVIXSI <= vixsi].idxmax()
            datos_test.loc[i,"QVIXSI"] = datos_train.loc[idq].QVIXSI.values[0]
        else:
            datos_test.loc[i,"QVIXSI"] = 0

        vix = datos_test.loc[i].FVIX
        if vix >= datos_train.FVIX.min():
            idq = datos_train[datos_train.FVIX < vix].idxmax()
            datos_test.loc[i,"QVIX"] = datos_train.loc[idq].QVIX.values[0]
        else:
            datos_test.loc[i,"QVIX"] = 0

    datos_train = datos_train[["QVIXSI","QVIX"]]
    datos_test = datos_test[["QVIXSI","QVIX"]]

    return datos_train, datos_test

datos_train, datos_test = datos_train_test (datos, "2008", "2019", "2020", "2024")
```

```python
def crear_secuencia (datos, LAG=11):

    datos_lag = np.zeros((datos.shape[0]-LAG,LAG,datos.shape[1]))

    for i in range(0, datos.shape[0]-LAG):
        datos_lag[i,:,:] = datos.iloc[i:i+LAG,:]

    return (datos_lag[:,0:-1,:], datos_lag[:,-1,0])

XX_tr, YY_tr = crear_secuencia (datos_train)
XX_ts, YY_ts = crear_secuencia (datos_test)
```

```python

def build_model_regresion(X):
    '''
    Se implementa el modelo de regresión lineal, con una sola capa densa de una neurona sin función de activación
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "REGRESION"
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(XX_tr.shape[1],XX_tr.shape[2])))
    model.add(keras.layers.Dense(1, activation=None))
    
    return model, model_name

def build_model_LSTM(X):
    '''
    Se implementa el modelo LSTM con una sola capa LSTM de tamaño temporal XX.shape[1] y canales XX.shape[2]
    La salida es una capa densa de una sola neurona sin función de activación
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''

    model_name = "LSTM"
    model = keras.Sequential()

    #return_sequences=True. Devuelve como salida toda la secuencia temporal en lugar de la última para la entrada de la siguiente capa
    #Como salida devuelve la misma dimensión temporal que la entrada

    model.add(keras.layers.LSTM(X.shape[1], input_shape=(X.shape[1],X.shape[2],), return_sequences=True))
    model.add(keras.layers.Flatten())
    model.add(Dense(1, activation=None))

    return model, model_name

def build_model_GRU(X):
    '''
    Se implementa el modelo GRU de la misma manera que el modelo LSTM
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "GRU"
    model = keras.Sequential()

    #return_sequences=True. Devuelve como salida toda la secuencia temporal en lugar de la última para la entrada de la siguiente capa
    #Como salida devuelve la misma dimensión temporal que la entrada

    model.add(keras.layers.GRU(X.shape[1], input_shape=(X.shape[1],X.shape[2],), return_sequences=True))
    model.add(keras.layers.Flatten())
    model.add(Dense(1, activation=None))

    return model, model_name

def build_model_CONV(X):
    '''
    Se implementa un modelo convolucional de dos capas de 16 y 8 filtros de tamaño 3 con pooling de 2 en cada capa para reducir la dimensión de las salidas
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''

    model_name = "CONV"
 
    model = keras.Sequential()
    model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='relu', input_shape=(X.shape[1],X.shape[2])))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Convolution1D(filters=8, kernel_size=3))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Flatten())
    model.add(Dense(1, activation=None))
              
    return model, model_name

def build_model_CONVLSTM(X):
    '''
    Se implementa un modelo convolucional y recurrente (lstm), donde la primera capa será un modelo convolicional que como salida genere la misma
    dimensión que la entrada para que pase por un modelo convolucional 1D similar al modelo anterior
    
    Se le pasa la variable de entrada X para recoger las dimensiones temporal y activos
    
    '''
    
    model_name = "CONVLSTM"

    model = keras.Sequential()
    
    #Le añade return_sequences=True para pasar toda la secuencia temporal al la capa convolucional
    model.add(keras.layers.LSTM(X.shape[1], input_shape=(X.shape[1],X.shape[2],), return_sequences=True))
    model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='relu', input_shape=(X.shape[1],X.shape[2])))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Convolution1D(filters=8, kernel_size=3))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Flatten())
    model.add(Dense(1, activation=None))
    
    return model, model_name

def build_model_attention (X):
    '''
    Se implementa un modelo de atención con tres capas densas (query, value y key) para transformar la salida de la capa LSTM en vectores de consulta,
    de valor y de clave como entrada en la capa attention
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "ATTENTION"

    XX = Input((X.shape[1],X.shape[2],))
    out = LSTM(10, return_sequences = True)(XX)

    query = Dense(X.shape[1])(out)
    value = Dense(X.shape[1])(out)
    key = Dense(X.shape[1])(out)

    out = Attention(use_scale=False)([query,value,key])
    out = Flatten()(out)
    out = Dense(1, activation=None)(out)

    model = Model(inputs=XX, outputs=out)
    
    return model, model_name

def build_model_bidireccional (X):
    '''
    Se implementa un modelo bidireccional asimétrico (GRU para forward yt LSTM para backward)
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "BIDIRECCIONAL"

    model = keras.Sequential()

    forward_layer = keras.layers.GRU(20,activation='relu',return_sequences=True)
    backward_layer = keras.layers.LSTM(20, activation='relu',return_sequences=True, go_backwards=True)

    model.add(keras.layers.Bidirectional(forward_layer, backward_layer=backward_layer,
                             input_shape=(X.shape[1],X.shape[2])))

    model.add(Flatten()) 
    model.add(Dense(1, activation=None))
    
    return model, model_name

    
    
def fit_eval_model (build_model, X, Y, Xt, Yt):
    '''
    
    Crea y entrena el modelo build_model con los datos de train X,Y
    y lo evalua con los datos de test Xt,Yt
    Saca la gráfica del entrenamiento 
    Devuelve los MSE de train, evaluación y test multiplicados por el factor de normalización maxdatos
    
    Si se introduce None en build_model, se predice con el valor anterior de la serie (modelo dummy)
    
    '''
    
    # Los datos de entrenamiento se desglosan en 80% entrenamiento y el 20% de evaluación
    Xtr, Xval, Ytr, Yval = train_test_split(X, Y, test_size=0.2, random_state=42)
    
    if build_model == None:
        model_name = "DUMMY"
        mse_tr =  np.mean(np.square(Ytr - np.roll(Ytr, shift=1)))
        mse_val =  np.mean(np.square(Yval - np.roll(Yval, shift=1)))
        mse_test =  np.mean(np.square(Yt - np.roll(Yt, shift=1)))

        print (f"MSE modelo {model_name} en datos de train: {mse_tr}, validación: {mse_val} y test: {mse_test}")

        return None, model_name, mse_tr, mse_val, mse_test

    else: 

        model, model_name = build_model(X)

        _learning_rate = 0.01
        _epochs = 100
        _batch_size = 32
        
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=_learning_rate), 
              loss=['mse'],
              metrics=['mse'])

        hist = model.fit(Xtr,
                         Ytr,
                         epochs=_epochs,
                         validation_data=(Xval, Yval),
                         batch_size=_batch_size,
                         verbose = 0)

        plt.figure(figsize=(10,6))
        plt.subplot(1,2,1)
        plt.plot(hist.history['loss'],label="loss")
        plt.plot(hist.history['val_loss'],label="val_loss")
        plt.legend()
        plt.subplot(1,2,2)
        plt.plot(hist.history['mse'],label="mse")
        plt.plot(hist.history['val_mse'],label="val_mse")
        plt.legend()

        _loss, mse_tr = model.evaluate(Xtr, Ytr, verbose=0)
        _loss, mse_val = model.evaluate(Xval, Yval, verbose=0)
        _loss, mse_test = model.evaluate(Xt, Yt, verbose=0)
        
        # los errores se multiplican por factor de normalización de datos
        mse_tr = mse_tr 
        mse_val = mse_val 
        mse_test = mse_test 

        print (f"MSE modelo {model_name} en datos de train: {mse_tr}, validación: {mse_val} y test: {mse_test}")

        return model, model_name, mse_tr, mse_val, mse_test
```
