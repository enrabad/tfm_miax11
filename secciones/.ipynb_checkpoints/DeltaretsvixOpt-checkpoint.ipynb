```python
# Se optimizan los precios con un delta operador para conseguir unos drawdowns objetivo
# y se generan los trades con dichos precios óptimos con el siguiente valor no visto siguiendo 
# con un walk forward de ventana expansiva con un único dato de test (el valor siguiente)


import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time
from pathlib import Path

def vix_investing ():
    '''
    
    '''
    
    url = "https://es.investing.com/indices/volatility-s-p-500-chart"
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
    else:
        print("no se ha podido cargar")
    soup = BeautifulSoup(html, "html.parser")
    vix_inv = soup.find('div', {'data-test': 'instrument-price-last'}).text
    fecha_inv = soup.find('time', {'data-test': 'trading-time-label'}).text
    vix_inv = float(vix_inv.replace(',', '.'))
    return vix_inv

def generar_vixsi_1m ():
    '''

    '''

    spy_yf = yf.download(tickers="SPY", interval="1m")
    ultima_fecha = spy_yf.index[-1]

    hora_exacta = ultima_fecha.time()

    df_vacio = True
    num_dias = 0
    while df_vacio:
      num_dias += 1
      if num_dias >= 15:
        print("no hay datos pasados de menos 15 días a hoy")
        return

      penultimo_dia_misma_hora = ultima_fecha - pd.Timedelta(days=num_dias)
      spy_ant = spy_yf.loc[
          (spy_yf.index.date == penultimo_dia_misma_hora.date()) & (spy_yf.index.time == hora_exacta)
      ]
      df_vacio = spy_ant.empty

    retspy = (spy_yf.iloc[-1].Close - spy_ant.iloc[-1].Close)/spy_ant.iloc[-1].Close

    vix_inv = vix_investing ()
    #fvixsi = np.abs(retspy) - vix_inv/(np.sqrt(256)*100)
    fvixsi = retspy - vix_inv/(np.sqrt(256)*100)
    vixsi = float(np.where ((fvixsi > 0), -1, 1))
    #20241005
    if retspy < 0:
        fvixsi = vix_inv/(np.sqrt(256)*100) - retspy

    #pd_vixsi1 = pd.DataFrame([vixsi], index=[ultima_fecha])
    #pd_vixsi1.columns = ["VIXSI"]
    #pd_vixsi = pd.concat([pd_vixsi, pd_vixsi1], axis=0)
    return (spy_yf.iloc[-1].Close, retspy, fvixsi)

def get_delta (datos, ret_activo, fvixsi):
    '''

    '''

    datos1 = datos.copy()
    datos1["retsig"] = datos1.rets.shift(-1)
    datos1.dropna(inplace=True)

    if ret_activo < 0:
        #if ret_activo > 0:
        #    datos1.NVIXSI = datos1.PVIXSI

        qnvixsi = datos1[datos1.rets < 0].NVIXSI.rank(pct=True)
        NVIXSI = pd.concat(
                    [qnvixsi, datos1.loc[qnvixsi.index].NVIXSI,
                     datos1.loc[qnvixsi.index].rets,
                     datos1.loc[qnvixsi.index].retsig],
                     axis=1)
        NVIXSI.columns = ["QVIXSI","FVIXSI","rets","retsig"]
#20241109 revision
        if fvixsi >= NVIXSI.FVIXSI.min():
            qvixsi = NVIXSI[NVIXSI.FVIXSI<fvixsi].QVIXSI.max()
        else:
            qvixsi = 0
        
    else:
        qpvixsi = datos1[datos1.rets > 0].PVIXSI.rank(pct=True)
        PVIXSI = pd.concat(
                    [qpvixsi, datos1.loc[qpvixsi.index].PVIXSI,
                     datos1.loc[qpvixsi.index].rets,
                     datos1.loc[qpvixsi.index].retsig],
                     axis=1)
        PVIXSI.columns = ["QVIXSI","FVIXSI","rets","retsig"]
#20241109 revision
        if fvixsi >= PVIXSI.FVIXSI.min():
            qvixsi = PVIXSI[PVIXSI.FVIXSI<fvixsi].QVIXSI.max()
        else:
            qvixsi = 0
#20241109 revision
    if ret_activo < 0:
        if qvixsi >= 0.5:
            # alcista
            qvixsi = (qvixsi - 0.5)*2
        else:
            # bajista
            qvixsi = (0.5 - qvixsi)*2 * -1
    else:
        if qvixsi >= 0.5:
            # bajista
            qvixsi = (qvixsi - 0.5)*2 * -1
        else:
            # alcista
            qvixsi = (0.5 - qvixsi)*2
    return qvixsi

#solo interesa Deltaindicators(Positivo,Negativo de 0 a 1, que indica la probabilidad de cambio)
def ini_deltavix (desde, hasta):
    '''
    
    '''

    fichero = f"datosretsvix/deltavix-{desde}_{hasta}.csv"
    fichero = Path(fichero)

    if fichero.is_file():
        #datos = pd.read_csv(f"datos2/deltavix-{desde}_{hasta}.csv", index_col=0, parse_dates=True)
        datos = pd.read_csv(fichero, index_col=0, parse_dates=True)
        return datos

    spy_datos = yf.download("SPY", start=desde, end=hasta)
    vix_datos = yf.download("^VIX", start=desde, end=hasta)
    #spx_datos = yf.download("^SPX", start=desde, end=hasta)
    #spx_datos = yf.download("SPY", start=desde, end=hasta)

    datos_all = pd.concat([spy_datos.Close, vix_datos.Close], axis=1).dropna()
    #datos_all.columns = ["SPY","VIX","SPX"]
    datos_all.columns = ["SPY","VIX"]
    datos_all["rets"] = datos_all.SPY.pct_change()
    #datos_all["xrets"] = datos_all.SPX.pct_change()
    datos_all["PVIXSI"] = datos_all.rets - datos_all.VIX/(np.sqrt(256)*100)
    datos_all["NVIXSI"] = datos_all.rets + datos_all.VIX/(np.sqrt(256)*100)
    #datos_all["FVIXSI"] = datos.rets - datos.VIX/(np.sqrt(256)*100)
    #datos_all["VIXSI"] = np.where ((datos.FVIXSI > 0), -1, 1)
    datos_all.dropna(inplace=True)
    #datos_all.to_csv(f"datos/deltavix-{desde}_{hasta}.csv")
    datos_all.to_csv(fichero)

    return datos_all

def get_deltavix (datos_all, desde_train, hasta_train):
    
    if desde_train == 0:
        fichero_train = f"datosretsvix/deltavix_{hasta_train}_train.csv"
    else:
        fichero_train = f"datosretsvixdeltavix_{desde_train}_{hasta_train}_train.csv"

    fichero_train = Path(fichero_train)
    if fichero_train.is_file():
        datos_train = pd.read_csv(fichero_train, index_col=0, parse_dates=True)
        return datos_train
    

    datos = datos_all.copy()
    datos = datos.iloc[desde_train:hasta_train]
    #datos["QVIXSI"] = datos.FVIXSI.rank(pct=True)
    datos.dropna(inplace=True)
    datos["qvixsi"] = -999
    
    ini = time.time()
    for i in range(len(datos)):
        #ret_activo = datos.iloc[i]["xrets"]
        ret_activo = datos.iloc[i]["rets"]
        if ret_activo > 0:
            fvixsi = datos.iloc[i]["PVIXSI"]
        else:
            fvixsi = datos.iloc[i]["NVIXSI"]

        qvixsi = get_delta(datos, ret_activo, fvixsi)
        datos.iloc[i,datos.columns.get_loc("qvixsi")] = qvixsi

    datos.to_csv(fichero_train)
    return datos

def rsharpe (srent):
    rent = srent.mean() * 252
    var = srent.std() * np.sqrt(252)
    rsharpe = rent/var
    eq_rent = srent.cumsum()
    drwd = (eq_rent - eq_rent.cummax())
    maxdrwd = drwd.min()
    return (rsharpe, maxdrwd)

def get_options ():
    df_op202401 = pd.read_csv("opciones/SPY2024-01.csv", index_col=0, parse_dates=True)
    df_op202402 = pd.read_csv("opciones/SPY2024-02.csv", index_col=0, parse_dates=True)
    df_op202403 = pd.read_csv("opciones/SPY2024-03.csv", index_col=0, parse_dates=True)
    df_op202404 = pd.read_csv("opciones/SPY2024-04.csv", index_col=0, parse_dates=True)
    df_op202405 = pd.read_csv("opciones/SPY2024-05.csv", index_col=0, parse_dates=True)
    df_op202406 = pd.read_csv("opciones/SPY2024-06.csv", index_col=0, parse_dates=True)
    df_op202407 = pd.read_csv("opciones/SPY2024-07.csv", index_col=0, parse_dates=True)
    df_op202408 = pd.read_csv("opciones/SPY2024-08.csv", index_col=0, parse_dates=True)
    df_op202409 = pd.read_csv("opciones/SPY2024-09.csv", index_col=0, parse_dates=True)
    df_op202410 = pd.read_csv("opciones/SPY2024-10.csv", index_col=0, parse_dates=True)
    df_op202411 = pd.read_csv("opciones/SPY2024-11.csv", index_col=0, parse_dates=True)
    df_op = pd.concat([df_op202401,
                      df_op202402,
                      df_op202403,
                      df_op202404,
                      df_op202405,
                      df_op202406,
                      df_op202407,
                      df_op202408,
                      df_op202409,
                      df_op202410,
                      df_op202411,                   
                      ], axis=0)
    
    df_op["fecha"] = pd.to_datetime(df_op["quote_date"])
    df_op["fexp"] = pd.to_datetime(df_op["expiration"])
    df_op["indice"] = df_op["fexp"]
    df_op = df_op[(df_op["quote_date"]!=df_op["expiration"])]
    df_op['delta'] = (df_op['fexp'] - df_op['fecha']).dt.days
    df_op.reset_index(inplace=True, drop=True)
    df_op.set_index('indice', inplace=True)
    df_op = df_op.loc[df_op.groupby('fecha')['delta'].idxmin()]

    df_op["CATM"] = df_op['underlying_ask_eod'].astype(int)
    df_op["resto"] = df_op['underlying_bid_eod']%1 > 0
    df_op["PATM"] = df_op['underlying_bid_eod'].astype(int) + df_op["resto"] * 1

    df_op["CSTRIKE"] = df_op.strike - df_op.CATM
    df_op["PSTRIKE"] = df_op.strike - df_op.PATM
    df_op = df_op[np.abs(df_op.CSTRIKE)<=50]
    df_op = df_op[["fecha","fexp","underlying_ask_eod",
                        "option_type","strike","bid_eod","ask_eod","CSTRIKE","PSTRIKE"]]
    df_op.columns=["FECHA","FEXP","ACTIVO","TIPO","STRIKE","BID","ASK","CSTRIKE","PSTRIKE"]
    
    df_opc = df_op.loc[(df_op.CSTRIKE == 0) & (df_op.TIPO=="C")].copy()
    df_opc['delta'] = (df_opc['FEXP'] - df_opc['FECHA']).dt.days
    df_opc.reset_index(inplace=True, drop=True)
    df_opc = df_opc.loc[df_opc.groupby('FECHA')['delta'].idxmin()]
    df_opc.set_index('FEXP', inplace=True)
    df_opc["cpuntos"] = df_opc["ACTIVO"].shift(-1) - df_opc["STRIKE"]
    df_opc["ccgain"] = np.where(df_opc.cpuntos > 0, df_opc.cpuntos - df_opc.ASK, - df_opc.ASK) 
    df_opc["vcgain"] = df_opc.ASK
    df_opc["vcgain"] = np.where(df_opc.cpuntos > 0, df_opc.ASK - df_opc.cpuntos, df_opc.ASK)   
    df_opc = df_opc[["ccgain", "vcgain", "cpuntos"]]
    
    df_opp = df_op.loc[(df_op.PSTRIKE == 0) & (df_op.TIPO=="P")].copy()
    df_opp['delta'] = (df_opp['FEXP'] - df_opp['FECHA']).dt.days
    df_opp.reset_index(inplace=True, drop=True)
    df_opp = df_opp.loc[df_opp.groupby('FECHA')['delta'].idxmin()]
    df_opp.set_index('FEXP', inplace=True)
    df_opp["ppuntos"] = df_opp["ACTIVO"].shift(-1) - df_opp["STRIKE"]
    df_opp["cpgain"] = np.where(df_opp.ppuntos < 0, - df_opp.ppuntos - df_opp.BID, - df_opp.BID) 
    df_opp["vpgain"] = df_opp.BID
    df_opp["vpgain"] = np.where(df_opp.ppuntos < 0, df_opp.BID + df_opp.ppuntos, df_opp.BID)   
    df_opp = df_opp[["cpgain", "vpgain", "ppuntos"]]
    
    return df_opc, df_opp
```

```python
df_opc, df_opp = get_options ()
```

```python
desde = "2008-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")
hasta = "2024-11-20"
datos = ini_deltavix (desde, hasta)
num_dias_test = len(datos.loc[df_opc.index[0]:])

ini = time.time()
df_qvixsi = pd.DataFrame()
for num_dias in range(num_dias_test):
    hasta_train = len(datos) - (num_dias_test - num_dias + 1)
    desde_train = 0
    
    datos_train = get_deltavix (datos, desde_train, hasta_train)
    
    retact = datos_train.iloc[-1]["rets"]
    if retact > 0:
        fvixsi = datos_train.iloc[-1]["PVIXSI"]
    else:
        fvixsi = datos_train.iloc[-1]["NVIXSI"]        

    qvixsi = get_delta(datos_train, retact, fvixsi)
    retsig = datos.iloc[hasta_train]["rets"]
    fechasig = datos.index[hasta_train]
    df_qvixsi1 = pd.DataFrame([[fechasig, qvixsi, retact, retsig]])
    df_qvixsi1.columns = ["fechasig", "qvixsi", "ret", "retsig"]
    df_qvixsi1.set_index("fechasig", inplace=True)
    df_qvixsi = pd.concat([df_qvixsi, df_qvixsi1], axis=0)
    
```

```python
df_op = pd.concat([df_opc, df_opp], axis=1)
trades = pd.concat([df_op, df_qvixsi], axis=1)
trades.dropna(inplace=True)
trades.to_csv("tradesretsvix.csv")
```

```python
def optimizar (trades):
    optmax2 = 10
    optmin2 = 0
    optmax1 = 10
    optmin1 = 0
    opt = 0

    for min1 in range(11):
        for max1 in range(11):
            for min2 in range(11):
                for max2 in range(11):
                    if max1 <= min1 or max2 <= min2:
                        continue

                    qvtrades = trades.copy()
                    qvtrades.loc[(qvtrades.qvixsi >= max1/10) |
                                 (qvtrades.qvixsi < -max2/10) ,"ccgain"] = 0
                    qvtrades.loc[(qvtrades.qvixsi < min1/10) & (qvtrades.qvixsi >= 0) |
                                 (qvtrades.qvixsi >= -min2/10) & (qvtrades.qvixsi < 0) ,"vcgain"] = 0
                    qvtrades.loc[(qvtrades.qvixsi >= max1/10) |
                                 (qvtrades.qvixsi < -max2/10) ,"vpgain"] = 0
                    qvtrades.loc[(qvtrades.qvixsi < min1/10) & (qvtrades.qvixsi >= 0) |
                                 (qvtrades.qvixsi >= -min2/10) & (qvtrades.qvixsi < 0) ,"cpgain"] = 0

                    opt_act = ((qvtrades.ccgain + qvtrades.cpgain).cumsum().iloc[-1] +
                              (qvtrades.vpgain + qvtrades.vcgain).cumsum().iloc[-1])
                    if opt_act > opt:
                        opt = opt_act
                        optmax1 = max1
                        optmin1 = min1
                        optmax2 = max2
                        optmin2 = min2
    return optmax1, optmin1, optmax2, optmin2


```

```python
# Optimización utilizando datos de train (el 80%) y datos de test (el 20% restante)

desde = trades.index[0]
hasta = trades.index[1]

num_dias_test = int(len(trades)*0.2)
datos_train = trades.iloc[0:-num_dias_test]
datos_test = trades.iloc[num_dias_test:]
optmax1, optmin1, optmax2, optmin2 = optimizar (datos_train)
qvtrades = datos_test
qtrades = qvtrades.loc[(qvtrades.qvixsi >= optmax1/10) |
                 (qvtrades.qvixsi < -optmax2/10) ,"ccgain"] = 0
    qvtrades.loc[(qvtrades.qvixsi < optmin1/10) & (qvtrades.qvixsi >= 0) |
                 (qvtrades.qvixsi >= -optmin2/10) & (qvtrades.qvixsi < 0) ,"vcgain"] = 0
    qvtrades.loc[(qvtrades.qvixsi >= optmax1/10) |
                 (qvtrades.qvixsi < -optmax2/10) ,"vpgain"] = 0
    qvtrades.loc[(qvtrades.qvixsi < optmin1/10) & (qvtrades.qvixsi >= 0) |
                 (qvtrades.qvixsi >= -optmin2/10) & (qvtrades.qvixsi < 0) ,"cpgain"] = 0
    
(qtrades.ccgain + qtrades.cpgain).cumsum().plot()
(qtrades.vpgain + qtrades.vcgain).cumsum().plot()
(qtrades.ccgain + qtrades.cpgain + qtrades.vpgain + qtrades.vcgain).cumsum().plot()
```
