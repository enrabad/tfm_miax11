### Creación cartera para actuar en corto en la estrategia 

Se ha visto en anteriores notebooks que el alpha generado por la extrategia VIXSI recae en la señal Vixsi -1, y que en la versión ponerse en corto, en promedio, cerca de la mitad de los activos del SP500 contienen un alfa superior al SPY, con lo que aparentemente no parece difícil encontrar una cartera de acciones que contenga un alfa superior al SPY (a partir de ahora, cuando se hable de alpha de un activo, siempre se referirá al alpha generado por la estrategia VIXSI en ese activo, que equivaldrá al doble de la rentabilidad de la señal en corto).

En este notebook se expone primero un ejemplo ideal, de una cartera formada únicamente por un activo con mucho alpha (KLAC), utilizando el activo en la posición corta de la estrategia VIXSI para el SPY, y se compara con la estrategia SPY.

Luego se intenta encontrar una cartera de K activos seleccionados de distintas maneras, utilizando como datos de entrada la diferencia de los alphas anuales entre los activos y el SPY.

Como los k activos y los métodos de selección son metaparámetros que se optimizarán, se utilizarán tres periodos para evaluar la cartera. Un periodo de training para encontrar los mejores activos a partir de unos metaparámetros fijados, un periodo de evaluación para encontrar los mejores metaparámetros, y un periodo de test, para probar la estrategia de la cartera en corto con los metaparámetros definitivos y SPY largo (a esta estrategia se le denominará a partir de ahora estrategia VIXSI mixta), y se compara con la estrategia VIXSI en SPY.

Al final se hará un resumen de los resultados obtenidos


```python
#Importamos las librerias necesarias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import yfinance as yf
import requests

%matplotlib inline
%load_ext autoreload
%autoreload 2
import warnings
warnings.filterwarnings('ignore')

from modulos.tfm_util import metrica_cartera
from modulos.tfm_util import drwdown
```

### Cargar señal vixi y alphas anualizados

```python
datos_vixsi = pd.read_csv ("../datos/vixsi.csv", index_col = 0)
datos_vixsi.index = pd.to_datetime(datos_vixsi.index)

df_alpha = pd.read_csv("../datos/alphas_ann.csv", index_col=0)
```

### Se cargan KLAC y SPY para crear la estrategia VIXSI mixta

```python
print (f"alpha medio anual KLAC: {df_alpha.KLAC.mean():.2f}%")
print (f"alpha medio anual SPY: {df_alpha.SPY.mean():.2f}%")
datos = yf.download(["KLAC","SPY"], datos_vixsi.index[0], datos_vixsi.index[-1] + pd.Timedelta(days=1))['Close']
```

### Se crea la equity de la estrategia VIXSI mixta con KLAC

```python
datos["vixsignal"] = datos_vixsi.VIXSI
datos["VIXSIMIXKLAC"] = (1 + ((datos.vixsignal.shift(1) + 1)/2)*datos.SPY.pct_change() +
                        ((datos.vixsignal.shift(1) - 1)/2)*datos.KLAC.pct_change()).cumprod()
datos["VIXSISPY"] = (1 + datos.vixsignal.shift(1) * datos.SPY.pct_change()).cumprod()
```

### Se analizan los resultados

La estrategia Vixsi mixta con KLAC mejora en todos los indicadores a la estrategia VIXSI SPY, ya que KLAC tiene un alpha medio que casi triplica al alpha de SPY. El objetivo será encontrar estos activos con algún método de selección de activos.

```python
estrategias = ["VIXSIMIXKLAC","VIXSISPY"]

ret_ann, ret_ann_b, r_sharpe, r_sharpe_b = metrica_cartera(datos[estrategias[0]], datos[estrategias[1]])
maxdwn, *_ = drwdown (datos[estrategias[0]])
maxdwn_b, *_ = drwdown (datos[estrategias[1]])

valores = [[ret_ann, r_sharpe, maxdwn], [ret_ann_b, r_sharpe_b, maxdwn_b]]
df_result = pd.DataFrame(valores, index= estrategias)
df_result.columns = ["RetAnual","RSharpe","MaxDwd"]

plt.figure(figsize=(10,6))
plt.plot(datos[estrategias[0]], label=estrategias[0])
plt.plot(datos[estrategias[1]], label=estrategias[1])

plt.legend()
plt.plot()
#plt.savefig('estrategia1.png')

df_result = df_result.round(2)
df_result.to_clipboard()
df_result
```

### Selección de activos para crear una estrategia vixsi mixta que mejore a la estrategia VIXSI SPY

Se va a selecciar los mejores activos en término de alpha desde el periodo de train-eval 2008:2018, utilizando el periodo  2008-2014 para el periodo train, donde se seleccionarán los mejores activos con un número de activos y un criterio de selección determinado, y se evaluarán los activos seleccionados para determinar los mejores metaparámetros (número de activos y criterios de selección) desde el 2015:2108.

Desde el 2019, se calcularán los mejores activos cada año, rolando anualmente los periodos de train-eval y test, para obtener la estrategia vixsi mixta en todo el periodo 2019-2024 y compararla con la estrategia vixsi en SPY.


### Se carga el dataset del SP500 y los alphas y se homogénean todos los datos

```python

data = pd.read_csv("../datos/sp500.csv", index_col=0)
data.index = pd.to_datetime(data.index)
indices = list(set(datos.index) & set(data.index))
data = data.loc[indices]
data.sort_index(inplace=True)

#solo interesan los retornos
data = data.pct_change()

df_alpha = pd.read_csv("../datos/alphas_ann.csv", index_col=0)
#Los alphas se restarán del alpha SPY, porque sólo interesa que sean superiores al SPY
df_alpha = df_alpha.sub(df_alpha.SPY, axis=0)

df_alpha = df_alpha[data.columns]
```

### Se inicializan los periodos 

### Se crean los diferentes metaparámetros: número de activos, funciones de selección de activos

```python


def mejor_dalpha_med (dalphas, k):
    '''
    Selecciona los k activos con la media de la diferencia de alphas entre los activos y SPY más altas
    
    '''

    return list(dalphas.mean().sort_values(ascending=False).index[:k])

def peor_dalpha_med (dalphas, k):
    '''
    Selecciona los k activos con la media de la diferencia de alphas entre los activos y SPY más bajas
    
    '''
        
    return list(dalphas.mean().sort_values(ascending=True).index[:k])

def mejor_dalpha_rsh (dalphas, k):
    '''
    Selecciona los k activos con el ratio media/var más altos
    
    '''
    
    return list((dalphas.mean()/dalphas.std()).sort_values(ascending=False).index[:k])

def peor_dalpha_rsh (dalphas, k):
    '''
    Selecciona los k activos con el ratio media/var más bajos
    
    '''
    return list((dalphas.mean()/dalphas.std()).sort_values(ascending=True).index[:k])

selfuncs = [mejor_dalpha_med, peor_dalpha_med, mejor_dalpha_rsh, peor_dalpha_rsh]
kactivos = [2,5,10]
```

```python
periodos_desde = [2008,0,2019]
periodos_hasta = [0,2018,2019]
df_periodos = pd.DataFrame([periodos_desde, periodos_hasta]).T
df_periodos.index = ["train","eval","test"]
df_periodos.columns = ["desde","hasta"]
for fechacorte_tr_ev in range (df_periodos.loc["eval", "hasta"] - df_periodos.loc["train", "desde"]):
    print(2008+corte_tr_ev, 2008+corte_tr_ev+1)
```

### Generación de la cartera

Se inicializa la estrategia y los periodos. Rolando cada año desde el 2019 al 2004:

1. Se seleccionarán los mejores activos para cada metaparámetro en el periodo de train

2. Se seleccionarán los parámetros con el mayor ratio alpha medio/desviación medio generado por los activos del punto 1 en el periodo de evaluación

3. Se genera la estrategia en el periodo de test 


```python

# Se inicializan valores de la estrategia
data["VIXSIMIXCART"] = 0
data["VIXSISPY"] = 0
data["vixsignal"] = datos.vixsignal.shift(1)
estrategias = ["VIXSIMIXCART","VIXSISPY"]

# Las fechas de corte entre train y eval será otro metaparámetro
periodos_desde = [2008,0,2019]
periodos_hasta = [0,2018,2019]
#periodos_desde = [2008,2013,2019]
#periodos_hasta = [2012,2018,2019]

df_periodos = pd.DataFrame([periodos_desde, periodos_hasta]).T
df_periodos.index = ["train","eval","test"]
df_periodos.columns = ["desde","hasta"]

for i in range(2024-df_periodos.loc["test","desde"] + 1):

    max_rdv = -999
    max_k = -1
    max_fsel = None
    max_activos = []

    for fechacorte_tr_ev in range (df_periodos.loc["eval", "hasta"] - df_periodos.loc["train", "desde"]):
        desde_train = str(df_periodos.loc["train", "desde"])
        #hasta_train = str(df_periodos.loc["train", "hasta"])
        #desde_eval = str(df_periodos.loc["eval", "desde"])
        hasta_train = fechacorte_tr_ev
        desde_eval = fechacorte_tr_ev + 1
        hasta_eval = str(df_periodos.loc["eval", "hasta"])
        desde_test = str(df_periodos.loc["test", "desde"])
        hasta_test = str(df_periodos.loc["test", "hasta"])
        tr_alpha = df_alpha.loc[desde_train:hasta_train]
        ev_alpha = df_alpha.loc[desde_eval:hasta_eval]

        for k in kactivos:
            for fsel in selfuncs:
                activos = fsel(tr_alpha, k)
                rdv = (ev_alpha[activos].mean()/ev_alpha[activos].std()).sum()/len(activos)
                if rdv > max_rdv:
                    max_rdv = rdv
                    max_k = k
                    max_fsel = fsel
                    max_activos = activos


    retornos = np.array (data.loc[desde_test:hasta_test,max_activos])
    señal_corto = np.array ((data.loc[desde_test:hasta_test].vixsignal-1)/2).reshape(-1,1)
    vixsimixto_corto = (retornos * señal_corto * 1/len(max_activos)).sum(axis=1)
    vixsimixto_largo = np.array((data.loc[desde_test:hasta_test].vixsignal+1)/2 * data.loc[desde_test:hasta_test].SPY)

    vixsimixto = vixsimixto_corto + vixsimixto_largo

    #tst_data["VIXSIMIXCART"] = pd.DataFrame((1 + pd.Series(vixsimixto)).cumprod()).set_index(tst_data.index)
    #tst_data["VIXSISPY"] = (1 + tst_data.vixsignal.shift(1) * tst_data.SPY).cumprod()
    data.loc[desde_test:hasta_test,estrategias[0]] = pd.DataFrame(pd.Series(vixsimixto)).set_index(
                                                    data.loc[desde_test:hasta_test].index)[0]
    data.loc[desde_test:hasta_test,estrategias[1]] = (data.loc[desde_test:hasta_test].vixsignal *
                                                    data.loc[desde_test:hasta_test].SPY)
    
    df_periodos = df_periodos + 1

data[estrategias] = (1 + data[estrategias]).cumprod()
```

### Test de la cartera

Comparación del alpha de la cartera combinada con el alpha del SPY periodo 2019 en adelante

```python

tdatos = data.loc["2019":,estrategias]
tdatos["VIXSIMIXCART"] = tdatos.VIXSIMIXCART/tdatos.VIXSIMIXCART.iloc[0]
tdatos["VIXSISPY"] = tdatos.VIXSISPY/tdatos.VIXSISPY.iloc[0]

ret_ann, ret_ann_b, r_sharpe, r_sharpe_b = metrica_cartera(tdatos[estrategias[0]], tdatos[estrategias[1]])
maxdwn, *_ = drwdown (tdatos[estrategias[0]])
maxdwn_b, *_ = drwdown (tdatos[estrategias[1]])

valores = [[ret_ann, r_sharpe, maxdwn], [ret_ann_b, r_sharpe_b, maxdwn_b]]
df_result = pd.DataFrame(valores, index= estrategias)
df_result.columns = ["RetAnual","RSharpe","MaxDwd"]

plt.figure(figsize=(10,6))
plt.plot(tdatos[estrategias[0]], label=estrategias[0])
plt.plot(tdatos[estrategias[1]], label=estrategias[1])

plt.legend()
plt.plot()
#plt.savefig('estrategia1.png')

df_result = df_result.round(2)
df_result.to_clipboard()
df_result
```
