# Optimización Walk Forward parámetro drawdown 
 
El walk forward se hará de dos maneras:

Con ventana expansiva de un periodo mínimo de 7 años para el entrenamiento y un periodo fijo de 1 año para el backtest. La ventana train|test va rolando de año en año, siendo la parte de train expansiva, es decir, siempre comienza desde el principio.

Con ventana fija de 7 años para el entrenamiento y un periodo fijo de 1 año para el backtest. La ventana train|test va rolando de año en año, siendo la parte de train de tamaño fijo.

Se podrían probar con otros periodos, en ese caso se tendría que utilizar un % del total del periodo para probar varios hiperparámetros y el resto de procentaje para evaluar los hiperparámetros seleccionados.

Los periodos de backtesting se concatenarán para analizar el backtesting de todo el periodo.

## Metaparámetros walk-forward

Se utiliza como año 256 días, que es el número de días bursátiles de un año. 
El periodo de train y test se asignan a 1 año


```python
class BestVixMomentum:
    '''
    
    '''
    
    def __init__(self, señal, num_señales=1):
        '''
        
        '''
        self.num_señales = num_señales
        self.señal = señal
        self._stops = [i for i in range(1,11)]
        #self._periodos = list((set(np.linspace(par_desde,par_hasta,num_par).astype(int))))
        self.stop_opt = None
        
    def crear_señal(self, datos, stop):
        '''


        '''

        vixmomsig = "vixmom_" + str(stop)
        maxdwn, dwserie = drwdown(datos.Close)
        datos[vixmomsig] = (dwserie > -0.01*stop)*1
        #datos[vixmomsig] = datos[vixmomsig] | (datos.Close.rolling(250).mean() > datos.Close)
        datos[vixmomsig]=datos[vixmomsig] + (1 - datos[vixmomsig])*datos[self.señal]

        return datos, vixmomsig

    def fit (self, datos):
        '''


        '''

        num_señales = self.num_señales
        periodos_df = pd.DataFrame()

        # Genera todas las posibles combinaciones de parámetros
        params = list(product(self._stops))
        
        # Muestra las combinaciones de parámetros
        for param in params:
            datos, señal = self.crear_señal(datos, param[0])
            metricas, _ = metricas_señal (datos, señal, comision=com)
            #(metricas["CAGR"]["Sharpe"]["MaxDrawdown"]
            periodos_df = pd.concat([periodos_df, pd.DataFrame([[param[0], metricas["CAGR"][0]]])], axis=0)

        periodos_df.columns = ["periodo","CAGR"]

        #print(periodos_df.sort_values(by="CAGR", ascending=False)[:num_señales])
        self.stops = int(periodos_df.sort_values(by="CAGR", ascending=False)[:num_señales].periodo.mean())

    def fit_eval (self, datos):
        '''


        '''
   
        datos, señal = self.crear_señal(datos, self.stops)
        return datos[señal]
```

<a id='Ejecutar_WF'></a> 
## Ejecutar walk forward ventana expansiva o ventana fija


```python
#0 se entrena en el primer periodo
#1 se entrena en cada periodo con ventana expansiva
#2 se entrena en cada periodo con ventana fija=periodo_train
modo_wf = 1
periodo_roll = 252
periodo_test = periodo_roll
periodo_train = periodo_test

señal = "vixsi"
com = 0.0005
datos_WF = pd.DataFrame(index=datos.index)
datos_WF = datos
metricas_wf = pd.DataFrame()
metricas_wfo = pd.DataFrame()
for i in range(periodo_train, len(datos), periodo_roll):
    señal_wf = "señal_WF"
    #señal_benchi="benchmark" + str(i)
    #señal_bench="benchmark"
    if i == periodo_train:
        datos_WF[señal_wf] = 0
    #datos_WF[señal_wfi] = 0
    
    desde = i - periodo_train
    hasta_train = i
    hasta_test = i + periodo_test
    ini = time.time()

    #bestvm = BestVixMomentum (1, par_hasta=np.where(periodo_train>252,252,int(periodo_train)/2), num_par=1)
    if modo_wf == 0 and i==periodo_train:
        bestvm = BestVixMomentum (señal)
        bestvm.fit(datos.iloc[:hasta_train])
    elif modo_wf == 1:
        bestvm = BestVixMomentum (señal)
        bestvm.fit(datos.iloc[:hasta_train])
    elif modo_wf == 2:
        bestvm = BestVixMomentum (señal)
        bestvm.fit(datos.iloc[desde:hasta_train])

    #datos_WF[señal_wfi].iloc[hasta_train:hasta_test] = bestvm.fit_eval(datos.iloc[hasta_train:hasta_test])
    datos_WF[señal_wf].iloc[hasta_train:hasta_test] = bestvm.fit_eval(datos.iloc[hasta_train:hasta_test])
    metrica_w, _ = metricas_señal(datos_WF.iloc[hasta_train:hasta_test], señal_wf , comision=com)
    i = len(metricas_wf)
    metricas_wf = pd.concat([metricas_wf, metrica_w], axis=0)
    metrica_v, _= metricas_señal(datos_WF.iloc[hasta_train:hasta_test], señal, plt_dwdown=False, comision=com)
    metricas_wf = pd.concat([metricas_wf, metrica_v], axis=0)
    metrica_b, _= metricas_señal(datos_WF.iloc[hasta_train:hasta_test], "benchmark", plt_dwdown=False)
    metricas_wf = pd.concat([metricas_wf, metrica_b], axis=0)
    j = len(metricas_wf)
    k = metricas_wf["CAGR"].iloc[i:j].argmax()
    estr = metricas_wf.index[k]
    desde = metricas_wf.iloc[i+k].desde
    hasta = metricas_wf.iloc[i+k].hasta
    metricas_wfo = pd.concat([metricas_wfo, pd.DataFrame([[estr,desde,hasta]])], axis=0)
    #print(modo_wfe, hasta_test, " tiempo:", time.time()-ini)


metricas_wfo.columns = ["Mayor CAGR","Desde","Hasta"]
```
