```python
def f_alpha(srets, brets):
    w_brets = brets.loc[srets.index]
    cov_rets = np.cov(srets.T, w_brets)[0,1:]
    var_brets = w_brets.var()
    beta = cov_rets / var_brets
    rent_srets = srets.sum()/len(srets)
    rent_brets = w_brets.sum()/len(w_brets)
    alpha = rent_srets - beta * rent_brets
    return (alpha)

def vixalpharoll(datos, periodo, activo):
    benchmark = "SPY"
    datos["vixsi"] = vixsignal(datos, benchmark=benchmark)
    ret_activos = np.log(datos[[benchmark]]).diff()
    ret_activos["vixsi"] = ret_activos[benchmark]*datos.vixsi.shift(1)
    alpha = ret_activos.rolling(periodo).apply(lambda x: f_alpha(x, ret_activos[benchmark]))
    alpha = alpha.fillna(0)
    return alpha["vixsi"]
```

```python
alphas = pd.DataFrame(index=datos.index)
alphas["SPY5"] = vixalpharoll(datos, periodo=5, activo="SPY")
alphas["SPY10"] = vixalpharoll(datos, periodo=10, activo="SPY")
alphas["SPY15"] = vixalpharoll(datos, periodo=15, activo="SPY")
alphas["SPY20"] = vixalpharoll(datos, periodo=20, activo="SPY")
alphas["SPY30"] = vixalpharoll(datos, periodo=30, activo="SPY")
alphas["SPY60"] = vixalpharoll(datos, periodo=60, activo="SPY")
alphas["SPY256"] = vixalpharoll(datos, periodo=256, activo="SPY")
```

```python
#0 se entrena en el primer periodo
#1 se entrena en cada periodo con ventana expansiva
#2 se entrena en cada periodo con ventana fija=periodo_train
modo_wf = 1
periodo_roll = 252
periodo_test = periodo_roll
periodo_train = periodo_test

svix="VIXM"
seal = "vixsi"
com = 0.0005
datos_WF = pd.DataFrame(index=datos.index)
#datos_WF[["Close","vixsi"]] = datos[["Close","vixsi"]]
datos_WF = datos
metricas_wf = pd.DataFrame()
metricas_wfo = pd.DataFrame()
for i in range(periodo_train, len(datos), periodo_roll):
    seal_wf = "seal_WF"
    #seal_benchi="benchmark" + str(i)
    #seal_bench="benchmark"
    if i == periodo_train:
        datos_WF[seal_wf] = 0
    #datos_WF[seal_wfi] = 0
    
    desde = i - periodo_train
    hasta_train = i
    hasta_test = i + periodo_test
    ini = time.time()

    #bestvm = BestVixMomentum (1, par_hasta=np.where(periodo_train>252,252,int(periodo_train)/2), num_par=1)
    if modo_wf == 0 and i==periodo_train:
        bestvm = BestVixMomentum (svix, seal)
        bestvm.fit(datos.iloc[:hasta_train])
    elif modo_wf == 1:
        bestvm = BestVixMomentum (svix, seal)
        bestvm.fit(datos.iloc[:hasta_train])
    elif modo_wf == 2:
        bestvm = BestVixMomentum (svix, seal)
        bestvm.fit(datos.iloc[desde:hasta_train])

    #datos_WF[seal_wfi].iloc[hasta_train:hasta_test] = bestvm.fit_eval(datos.iloc[hasta_train:hasta_test])
    datos_WF[seal_wf].iloc[hasta_train:hasta_test] = bestvm.fit_eval(datos.iloc[hasta_train:hasta_test])
    metrica_w, _ = metricas_vix(datos_WF.iloc[hasta_train:hasta_test], svix, seal_wf , comision=com)
    i = len(metricas_wf)
    metricas_wf = pd.concat([metricas_wf, metrica_w], axis=0)
    metrica_v, _= metricas_vix(datos_WF.iloc[hasta_train:hasta_test], svix, seal, plt_dwdown=False, comision=com)
    metricas_wf = pd.concat([metricas_wf, metrica_v], axis=0)
    metrica_b, _= metricas_vix(datos_WF.iloc[hasta_train:hasta_test], svix, "benchmark", plt_dwdown=False)
    metricas_wf = pd.concat([metricas_wf, metrica_b], axis=0)
    j = len(metricas_wf)
    k = metricas_wf["CAGR"].iloc[i:j].argmax()
    estr = metricas_wf.index[k]
    desde = metricas_wf.iloc[i+k].desde
    hasta = metricas_wf.iloc[i+k].hasta
    metricas_wfo = pd.concat([metricas_wfo, pd.DataFrame([[estr,desde,hasta]])], axis=0)
    #print(modo_wfe, hasta_test, " tiempo:", time.time()-ini)


metricas_wfo.columns = ["Mayor CAGR","Desde","Hasta"]
```
