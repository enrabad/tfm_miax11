```python
#El objetivo de este indicador es ayudar a crear estrategias credit y debit spreads tanto para call como para put


import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time
from pathlib import Path

def vix_investing ():
    '''
    
    '''
    
    url = "https://es.investing.com/indices/volatility-s-p-500-chart"
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
    else:
        print("no se ha podido cargar")
    soup = BeautifulSoup(html, "html.parser")
    vix_inv = soup.find('div', {'data-test': 'instrument-price-last'}).text
    fecha_inv = soup.find('time', {'data-test': 'trading-time-label'}).text
    vix_inv = float(vix_inv.replace(',', '.'))
    return vix_inv

def generar_vixsi_1m ():
    '''

    '''

    spy_yf = yf.download(tickers="SPY", interval="1m")
    ultima_fecha = spy_yf.index[-1]

    hora_exacta = ultima_fecha.time()

    df_vacio = True
    num_dias = 0
    while df_vacio:
      num_dias += 1
      if num_dias >= 15:
        print("no hay datos pasados de menos 15 dÃ­as a hoy")
        return

      penultimo_dia_misma_hora = ultima_fecha - pd.Timedelta(days=num_dias)
      spy_ant = spy_yf.loc[
          (spy_yf.index.date == penultimo_dia_misma_hora.date()) & (spy_yf.index.time == hora_exacta)
      ]
      df_vacio = spy_ant.empty

    retspy = (spy_yf.iloc[-1].Close - spy_ant.iloc[-1].Close)/spy_ant.iloc[-1].Close

    vix_inv = vix_investing ()
    #fvixsi = np.abs(retspy) - vix_inv/(np.sqrt(256)*100)
    fvixsi = retspy - vix_inv/(np.sqrt(256)*100)
    vixsi = float(np.where ((fvixsi > 0), -1, 1))
    #20241005
    if retspy < 0:
        fvixsi = vix_inv/(np.sqrt(256)*100) - retspy

    #pd_vixsi1 = pd.DataFrame([vixsi], index=[ultima_fecha])
    #pd_vixsi1.columns = ["VIXSI"]
    #pd_vixsi = pd.concat([pd_vixsi, pd_vixsi1], axis=0)
    return (spy_yf.iloc[-1].Close, retspy, fvixsi)

def get_delta(datos, ret_activo, fvixsi):
    '''

    '''

    datos1 = datos.copy()
    datos1["xretsig"] = datos1.xrets.shift(-1)
    datos1.dropna(inplace=True)

    if ret_activo < 0:
        #if ret_activo > 0:
        #    datos1.NVIXSI = datos1.PVIXSI

        qnvixsi = datos1[datos1.xrets < 0].NVIXSI.rank(pct=True)
        NVIXSI = pd.concat(
                    [qnvixsi, datos1.loc[qnvixsi.index].NVIXSI,
                     datos1.loc[qnvixsi.index].xrets,
                     datos1.loc[qnvixsi.index].xretsig],
                     axis=1)
        NVIXSI.columns = ["QVIXSI","FVIXSI","xrets","xretsig"]
#20241109 revision
        if fvixsi >= NVIXSI.FVIXSI.min():
            qvixsi = NVIXSI[NVIXSI.FVIXSI<fvixsi].QVIXSI.max()
        else:
            qvixsi = 0
        
    else:
        qpvixsi = datos1[datos1.xrets > 0].PVIXSI.rank(pct=True)
        PVIXSI = pd.concat(
                    [qpvixsi, datos1.loc[qpvixsi.index].PVIXSI,
                     datos1.loc[qpvixsi.index].xrets,
                     datos1.loc[qpvixsi.index].xretsig],
                     axis=1)
        PVIXSI.columns = ["QVIXSI","FVIXSI","xrets","xretsig"]
#20241109 revision
        if fvixsi >= PVIXSI.FVIXSI.min():
            qvixsi = PVIXSI[PVIXSI.FVIXSI<fvixsi].QVIXSI.max()
        else:
            qvixsi = 0
#20241109 revision
    if ret_activo < 0:
        if qvixsi >= 0.5:
            # alcista
            qvixsi = (qvixsi - 0.5)*2
        else:
            # bajista
            qvixsi = (0.5 - qvixsi)*2 * -1
    else:
        if qvixsi >= 0.5:
            # bajista
            qvixsi = (qvixsi - 0.5)*2 * -1
        else:
            # alcista
            qvixsi = (0.5 - qvixsi)*2
    return qvixsi

#solo interesa Deltaindicators(Positivo,Negativo de 0 a 1, que indica la probabilidad de cambio)
def ini_deltavix (desde, hasta, desde_train, hasta_train):
    '''
    
    '''

    fichero = f"deltavix-{desde}_{hasta}.csv"
    fichero = Path(fichero)
    fichero_train = f"deltavix-{desde_train}_{hasta_train}_train.csv"
    fichero_train = Path(fichero_train)

    if fichero.is_file() and fichero_train.is_file():
        datos = pd.read_csv(f"deltavix-{desde}_{hasta}.csv", index_col=0, parse_dates=True)
        datos_train = pd.read_csv(f"deltavix-{desde_train}_{hasta_train}_train.csv", index_col=0, parse_dates=True)
        return datos, datos_train

    spy_datos = yf.download("SPY", start=desde, end=hasta)
    vix_datos = yf.download("^VIX", start=desde, end=hasta)
    spx_datos = yf.download("^SPX", start=desde, end=hasta)
    #spx_datos = yf.download("SPY", start=desde, end=hasta)

    datos_all = pd.concat([spy_datos.Close, vix_datos.Close, spx_datos.Close], axis=1).dropna()
    datos_all.columns = ["SPY","VIX","SPX"]
    datos_all["rets"] = datos_all.SPY.pct_change()
    datos_all["xrets"] = datos_all.SPX.pct_change()
    datos_all["PVIXSI"] = datos_all.rets - datos_all.VIX/(np.sqrt(256)*100)
    datos_all["NVIXSI"] = datos_all.rets + datos_all.VIX/(np.sqrt(256)*100)
    #datos_all["FVIXSI"] = datos.rets - datos.VIX/(np.sqrt(256)*100)
    #datos_all["VIXSI"] = np.where ((datos.FVIXSI > 0), -1, 1) 
    datos = datos_all.copy()
    datos = datos_all.loc[desde_train:hasta_train]
    #datos["QVIXSI"] = datos.FVIXSI.rank(pct=True)
    datos.dropna(inplace=True)
    datos["qvixsi"] = -999
    
    ini = time.time()
    for i in range(len(datos)):
        ret_activo = datos.iloc[i]["xrets"]
        if ret_activo > 0:
            fvixsi = datos.iloc[i]["PVIXSI"]
        else:
            fvixsi = datos.iloc[i]["NVIXSI"]

        qvixsi = get_delta(datos, ret_activo, fvixsi)
#20241109 revision
        '''
        if ret_activo < 0:
            if qvixsi >= 0.5:
                # alcista
                qvixsi = (qvixsi - 0.5)*2
            else:
                # bajista
                qvixsi = (0.5 - qvixsi)*2 * -1
        else:
            if qvixsi >= 0.5:
                # bajista
                qvixsi = (qvixsi - 0.5)*2 * -1
            else:
                # alcista
                qvixsi = (0.5 - qvixsi)*2
        '''

        datos.iloc[i,datos.columns.get_loc("qvixsi")] = qvixsi

    #datos_all.to_csv(f"deltavix-{desde}_{hasta}.csv")
    #datos.to_csv(f"deltavix-{desde_train}_{hasta_train}_train.csv")
    return datos_all, datos

def generar_datos_deltavix (datos, activo, ret_activo, fvixsi):

    #datos = generar_deltavix (desde, hasta)
    #datos = generar_deltavix_file ()

    sl = 0.01
    fvixsi = get_delta(datos, ret_activo, fvixsi)
    for i in range (10):
        j = i + 1
        if fvixsi == 0 or fvixsi > i/10 and fvixsi <= j/10:
            PQrets1 = datos[(datos.qvixsi.shift(1)>i/10)&(datos.qvixsi.shift(1)<=j/10)][["SPX","xrets"]]
            slp = -sl
            sln = 100
            sln = sl
        elif fvixsi < -i/10 and fvixsi >= -j/10:
            PQrets1 = datos[(datos.qvixsi.shift(1)<-i/10)&(datos.qvixsi.shift(1)>=-j/10)][["SPX","xrets"]]
            slp = 100
            slp = -sl
            sln = sl

    deltavix = pd.DataFrame(index = [k for k in range(-35,40,5)])
    deltavix["PutDelta"]=0
    deltavix["PutPrecio"]=0
    deltavix["PutVentaPrecio"]=0
    deltavix["PutCompraPrecio"]=0
    deltavix["CallDelta"]=0
    deltavix["CallPrecio"]=0
    deltavix["CallCompraPrecio"]=0
    deltavix["CallVentaPrecio"]=0

    if (sl*datos.SPX.iloc[-1]%5 > 2):
        spread = (int(sl*datos.SPX.iloc[-1]/5)+1)*5
    else:
        spread = {int(sl*datos.SPX.iloc[-1]/5)*5}
    # deltavix["sl"] = f"deltavol {fvixsi} cierre spread a {spread}"
    deltavix["sl"] = sl
    deltavix["qvixsi"] = fvixsi

    # La esperanza matematica se asegura con la media mÃ¡s/menos la varianza de la serie PQrets1
    for k in range(-35,40,5):
#20241109 revision
        #if datos.SPX.iloc[-1]%5 == 0:
        if activo % 5 == 0:
            #max = k/datos.SPX.iloc[-1]
            max = k/activo
        else:
            #max = (k + (5 - datos.SPX.iloc[-1]%5))/datos.SPX.iloc[-1]
            max = (k + (5 - activo % 5))/activo
        #max = k/datos.SPX.iloc[-1]
        PQrets1["rets"] = 0
        PQrets1.loc[PQrets1.xrets<max,"rets"] = PQrets1[PQrets1.xrets<max].xrets - max
        PQrets1.loc[PQrets1.xrets<slp,"rets"] = slp
        #preciov = np.abs((PQrets1.rets.mean() - 2*np.abs(PQrets1.rets.var()))  * datos.SPX.iloc[-1])
        #precioc = np.abs((PQrets1.rets.mean() + 2*np.abs(PQrets1.rets.var()))  * datos.SPX.iloc[-1])
        #precio = np.abs(PQrets1.rets.mean() * datos.SPX.iloc[-1])
#20241109 revision
        #preciov = np.abs((PQrets1.rets.mean() - 2*np.abs(PQrets1.rets.var()))  * activo)
        preciov = np.abs((PQrets1.rets.mean() + 2*np.abs(PQrets1.rets.var()))  * activo)
        precioc = np.abs((PQrets1.rets.mean() - 2*np.abs(PQrets1.rets.var()))  * activo)
        precio = np.abs(PQrets1.rets.mean() * activo)
        delta = len(PQrets1.loc[PQrets1.xrets<max])/len(PQrets1)
        deltavix.loc[k,"PutPrecio"] = precio
        deltavix.loc[k,"PutVentaPrecio"] = preciov
        deltavix.loc[k,"PutCompraPrecio"] = precioc
        deltavix.loc[k,"PutDelta"] = delta
    for k in range(-35,40,5):
#20241109 revision
        #if datos.SPX.iloc[-1]%5 == 0:
        if activo % 5 == 0:
            #max = k/datos.SPX.iloc[-1]
            max = k/activo
        else:
            #max = (k - datos.SPX.iloc[-1]%5)/datos.SPX.iloc[-1]
            max = (k - activo % 5)/activo
        #max = k/datos.SPX.iloc[-1]
        PQrets1["rets"] = 0
        PQrets1.loc[PQrets1.xrets>max,"rets"] = max - PQrets1[PQrets1.xrets>max].xrets
        PQrets1.loc[PQrets1.xrets>sln,"rets"] = -sln
        #preciov = np.abs((PQrets1.rets.mean() - 2*PQrets1.rets.var())  * datos.SPX.iloc[-1])
        #precioc = np.abs((PQrets1.rets.mean() + 2*PQrets1.rets.var())  * datos.SPX.iloc[-1])
        #precio = np.abs(PQrets1.mean().rets * datos.SPX.iloc[-1])
#20241109 revision
        preciov = np.abs((PQrets1.rets.mean() + 2*PQrets1.rets.var())  * activo)
        precioc = np.abs((PQrets1.rets.mean() - 2*PQrets1.rets.var())  * activo)
        precio = np.abs(PQrets1.mean().rets * activo)
        delta = len(PQrets1.loc[PQrets1.xrets>max])/len(PQrets1)
        deltavix.loc[k,"CallPrecio"] = precio
        deltavix.loc[k,"CallVentaPrecio"] = preciov
        deltavix.loc[k,"CallCompraPrecio"] = precioc
        deltavix.loc[k,"CallDelta"] = delta

#############################################

    return deltavix, PQrets1
```

```python
def rsharpe (srent):
    rent = srent.mean() * 252
    var = srent.std() * np.sqrt(252)
    rsharpe = rent/var
    eq_rent = srent.cumsum()
    drwd = (eq_rent - eq_rent.cummax())
    maxdrwd = drwd.min()
    return (rsharpe, maxdrwd)

desde = "2008-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")
desde_train = desde
hasta_train = hasta
datos, datos_train = ini_deltavix (desde, hasta, desde_train, hasta_train)

#VENTA DE PUT
ventaput = pd.DataFrame()
v_sl = 0.01
v_maxdd = -0.04
v_rshp = 1
for k in range(-35,40,5):
    for i in range(-10,10):
            minmax_delta = 500
            qrets = datos_train[(datos_train.qvixsi.shift(1)>i/10)&(datos_train.qvixsi.shift(1)<=(i+1)/10)][["SPX","xrets"]]
            rmax = (2 + k)/qrets.SPX[-1]
            qrets["rets"] = 0
            qrets.loc[qrets.xrets<max,"rets"] = qrets[qrets.xrets<max].xrets - max
            qrets.loc[qrets.xrets<-v_sl,"rets"] = -v_sl - max

            for delta in range(500):
                precio = np.abs(qrets.rets.mean()) + qrets.rets.var() * delta
                qrets["rets_obj"] = qrets.rets + precio
                rshp, maxdw = rsharpe(qrets.rets_obj)
                if maxdw > v_maxdd and rshp > v_rshp:
                    minmax_dd = maxdw
                    minmax_i = i
                    minmax_j = i + 1
                    minmax_delta = delta
                    minmax_precio = precio
                    minmax_rs = rshp
                    minmax_qrets = qrets.rets_obj
                    break
            if minmax_delta < 500:
                deltaput = ["ventaput",k, minmax_i/10, minmax_j/10, minmax_delta, minmax_precio, minmax_rs, minmax_dd]
                #print ("k:", k, "i:", minmax_i, "j:", minmax_i + 1, "delta:",minmax_delta,"rsharpe:",minmax_rs,"maxddown:",minmax_dd)
                ventaput = pd.concat([ventaput, pd.DataFrame([deltaput])], axis=0)
            else:
                print ("k:", k, " no tiene soluciÃ³n", i, i+1)
                
ventaput.columns = ["operacion", "strike", "d_qvixsi", "h_qvixsi", "delta", "precio", "rsharpe","dwdown"]
ventaput.reset_index(inplace=True,drop=True)
```
