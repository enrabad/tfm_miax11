```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time

#df_vixsi = pd.read_csv("/content/drive/MyDrive/vixsi/VIXSIyf.csv", index_col=0)
#df_vixsi.index =  pd.to_datetime(df_vixsi.index)
#desde = df_vixsi.index[0].strftime("%Y-%m-%d")
#hasta = df_vixsi.index[-1].strftime("%Y-%m-%d")

desde = "2008-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")

spy_datos = yf.download("SPY", start=desde, end=hasta)
vix_datos = yf.download("^VIX", start=desde, end=hasta)

datos = pd.concat([spy_datos.Close, vix_datos.Close], axis=1).dropna()
datos.columns = ["SPY","VIX"]
datos.dropna(inplace=True)
datoS = datos.resample('W').last()
datoS.columns = ["SPYS", "VIXS"]
datoS["QVIXS"] = datoS.VIXS.rank(pct=True)
```

```python
datoS["retSPY"] = np.log(datoS.SPYS).diff()
datoS["señal"] = 0
datoS.loc[datoS.QVIXS<0.9,"señal"] = 1
(datoS.señal * datoS.retSPY.shift(-1)).cumsum().plot()
```

```python
datos["QVIXS"] = datos.VIX.rank(pct=True)
datos["retSPY"] = np.log(datos.SPY).diff()
datos["señal"] = 0
datos.loc[datos.QVIXS<0.8,"señal"] = 1
(datos.señal * datos.retSPY.shift(-1)).cumsum().plot()
```
