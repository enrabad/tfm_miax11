```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time

#df_vixsi = pd.read_csv("/content/drive/MyDrive/vixsi/VIXSIyf.csv", index_col=0)
#df_vixsi.index =  pd.to_datetime(df_vixsi.index)
#desde = df_vixsi.index[0].strftime("%Y-%m-%d")
#hasta = df_vixsi.index[-1].strftime("%Y-%m-%d")


def vix_investing ():
    '''
    
    '''
    
    url = "https://es.investing.com/indices/volatility-s-p-500-chart"
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
    else:
        print("no se ha podido cargar")
    soup = BeautifulSoup(html, "html.parser")
    vix_inv = soup.find('div', {'data-test': 'instrument-price-last'}).text
    fecha_inv = soup.find('time', {'data-test': 'trading-time-label'}).text
    vix_inv = float(vix_inv.replace(',', '.'))
    return vix_inv

def generar_vixsi_1m ():
    '''
    
    '''

    spy_yf = yf.download(tickers="SPY", interval="1m")
    ultima_fecha = spy_yf.index[-1]
    
    pd_vixsi = pd.DataFrame()

    hora_exacta = ultima_fecha.time()
    
    df_vacio = True
    num_dias = 0
    while df_vacio:
      num_dias += 1
      if num_dias >= 15:
        print("no hay datos pasados de menos 15 dÃ­as a hoy")
        return

      penultimo_dia_misma_hora = ultima_fecha - pd.Timedelta(days=num_dias)
      spy_ant = spy_yf.loc[
          (spy_yf.index.date == penultimo_dia_misma_hora.date()) & (spy_yf.index.time == hora_exacta)
      ]
      df_vacio = spy_ant.empty

    retspy = (spy_yf.iloc[-1].Close - spy_ant.iloc[-1].Close)/spy_ant.iloc[-1].Close

    vix_inv = vix_investing ()
    #fvixsi = np.abs(retspy) - vix_inv/(np.sqrt(256)*100)
    fvixsi = retspy - vix_inv/(np.sqrt(256)*100)
    vixsi = np.float(np.where ((fvixsi > 0), -1, 1))
    #20241005
    if retspy < 0:
        fvixsi = vix_inv/(np.sqrt(256)*100) - retspy
        
    #pd_vixsi1 = pd.DataFrame([vixsi], index=[ultima_fecha])
    #pd_vixsi1.columns = ["VIXSI"]
    #pd_vixsi = pd.concat([pd_vixsi, pd_vixsi1], axis=0)
    return (retspy, fvixsi)

desde = "2008-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")

spy_datos = yf.download("SPY", start=desde, end=hasta)
vix_datos = yf.download("^VIX", start=desde, end=hasta)
spx_datos = yf.download("^SPX", start=desde, end=hasta)

datos = pd.concat([spy_datos.Close, vix_datos.Close, spx_datos.Close], axis=1).dropna()
datos.columns = ["SPY","VIX","SPX"]
datos["rets"] = datos.SPY.pct_change()
datos["xrets"] = datos.SPX.pct_change()
#datos["FVIXSI"] = np.where (datos.rets > 0, datos.rets - datos.VIX/(np.sqrt(256)*100), - datos.VIX/(np.sqrt(256)*100))
datos["FVIXSI"] = datos.rets - datos.VIX/(np.sqrt(256)*100)
#20241005
datos["PVIXSI"] = datos.rets - datos.VIX/(np.sqrt(256)*100)
datos["NVIXSI"] = datos.rets + datos.VIX/(np.sqrt(256)*100)
datos["VIXSI"] = np.where ((datos.FVIXSI > 0), -1, 1)
datos["QVIXSI"] = datos.FVIXSI.rank(pct=True)
datos.dropna(inplace=True)
activo = datos.iloc[-1].SPX

def get_delta(datos, ret_activo, fvixsi, op, porce_atm, alcista):
    '''
    
    '''
    
    datos1 = datos.copy()
    datos1["xretsig"] = datos1.xrets.shift(-1)
    datos1.dropna(inplace=True)

    if ret_activo < 0:
        #if ret_activo > 0:
        #    datos1.NVIXSI = datos1.PVIXSI

        qnvixsi = datos1[datos1.xrets < 0].NVIXSI.rank(pct=True)
        NVIXSI = pd.concat(
                    [qnvixsi, datos1.loc[qnvixsi.index].NVIXSI,
                     datos1.loc[qnvixsi.index].xrets,
                     datos1.loc[qnvixsi.index].xretsig],
                     axis=1)
        NVIXSI.columns = ["QVIXSI","FVIXSI","xrets","xretsig"]

        qvixsi = NVIXSI[NVIXSI.FVIXSI<fvixsi].QVIXSI.max()
            #datos2 = datos1[(datos1.QVIXSI<qvixsi)]
        if alcista:
            datos2 = datos1[(datos1.NVIXSI>=fvixsi)]
        else:
            datos2 = datos1[(datos1.NVIXSI<=fvixsi)]
        if op == "P":
            #datos2 = datos1[(datos1.NVIXSI>fvixsi)]
            return len(datos2[datos2.xretsig < porce_atm])/len(datos2)*-1, qvixsi
        else:
            #datos2 = datos1[(datos1.NVIXSI>fvixsi)]
            return len(datos2[datos2.xretsig > porce_atm])/len(datos2), qvixsi
    else:
        #if ret_activo < 0:
        #    datos1.PVIXSI = datos1.NVIXSI

        qpvixsi = datos1[datos1.xrets > 0].PVIXSI.rank(pct=True)
        PVIXSI = pd.concat(
                    [qpvixsi, datos1.loc[qpvixsi.index].PVIXSI,
                     datos1.loc[qpvixsi.index].xrets,
                     datos1.loc[qpvixsi.index].xretsig],
                     axis=1)
        PVIXSI.columns = ["QVIXSI","FVIXSI","xrets","xretsig"]

        qvixsi = PVIXSI[PVIXSI.FVIXSI<fvixsi].QVIXSI.max()
        
        if alcista:
            datos2 = datos1[(datos1.PVIXSI<=fvixsi)]
        else:
            datos2 = datos1[(datos1.PVIXSI>=fvixsi)]

        if op == "P":
            #datos2 = PVIXSI[(datos1.PVIXSI>fvixsi)]
            return len(datos2[datos2.xretsig < porce_atm])/len(datos2)*-1, qvixsi
        else:
            #datos2 = datos1[(datos1.PVIXSI>fvixsi)]
            return len(datos2[datos2.xretsig > porce_atm])/len(datos2), qvixsi

def generar_deltavixsi (datos, activo, alcista):
    ret_activo, fvixsi = generar_vixsi_1m ()
    #if ret_activo > 0:
    #    alcista = True
    #else:
    #    alcista = False
    #fvixsi = 0.03
    #fvixsi = 0.05
    p_deltavix = pd.DataFrame()
    c_deltavix = pd.DataFrame()
    op="P"
    for i in range(-50,55,5):
        strike = (activo + activo % 5) + i
        porce_atm = (strike - activo)/activo
        _deltavix, qvixsi = get_delta(datos, ret_activo, fvixsi, op, porce_atm, alcista)
        #print(f"posicion:{i} strike:{strike} %atm:{porce_atm:.3f} deltavix:{deltavix:.3f}")
        ind = i
        deltavix = pd.DataFrame([_deltavix], index=[ind])
        if alcista:
            deltavix.columns = ["PUT_DELTAVIX"]
        else:
            deltavix.columns = ["PUT_DELTAVIX"]
        p_deltavix = pd.concat([p_deltavix, deltavix], axis=0)

    op="C"
    for i in range(-50,55,5):
        strike = (activo - activo % 5) + i
        porce_atm = (strike - activo)/activo
        _deltavix, qvixsi = get_delta(datos, ret_activo, fvixsi, op, porce_atm, alcista)
        ind = i
        deltavix = pd.DataFrame([_deltavix], index=[ind])
        if alcista:
            deltavix.columns = ["CALL_DELTAVIX"]
        else:
            deltavix.columns = ["CALL_DELTAVIX"]
        c_deltavix = pd.concat([c_deltavix, deltavix], axis=0)

    pd_deltavix = pd.concat([c_deltavix, p_deltavix], axis=1)
    pd_deltavix.index.name = "OFFATM"
    return c_deltavix, p_deltavix , qvixsi, fvixsi, ret_activo

c_deltavixa, p_deltavixa, qvixsi, fvixsi, ret_activo = generar_deltavixsi (datos, activo, alcista=True)
print (f"DELTAVIX ALCISTA {qvixsi:.3f} fvixsi: {fvixsi:.5f} RET: {ret_activo:.5f}")

c_deltavixb, p_deltavixb, qvixsi, fvixsi, ret_activo = generar_deltavixsi (datos, activo, alcista=False)
print (f"DELTAVIX BAJISTA {qvixsi:.3f} fvixsi: {fvixsi:.5f} RET: {ret_activo:.5f}")

if ret_activo < 0:
    if qvixsi >= 0.5:
        qvixsi = (qvixsi - 0.5)*2
        print (f"VIXSIGNAL ALCISTA {qvixsi:.2f}")
        pd_deltavix = pd.concat([c_deltavixa, p_deltavixa], axis=1)
        pd_deltavix.index.name = "OFFATM"
    else:
        qvixsi = (0.5 - qvixsi)*2
        print (f"VIXSIGNAL BAJISTA {qvixsi:.2f}")
        pd_deltavix = pd.concat([c_deltavixb, p_deltavixb], axis=1)
        pd_deltavix.index.name = "OFFATM"
else:
    if qvixsi >= 0.5:
        qvixsi = (qvixsi - 0.5)*2
        print (f"VIXSIGNAL BAJISTA {qvixsi:.2f}")
        pd_deltavix = pd.concat([c_deltavixb, p_deltavixb], axis=1)
        pd_deltavix.index.name = "OFFATM"
    else:
        qvixsi = (0.5 - qvixsi)*2
        print (f"VIXSIGNAL ALCISTA {qvixsi:.2f}")
        pd_deltavix = pd.concat([c_deltavixa, p_deltavixa], axis=1)
        pd_deltavix.index.name = "OFFATM"

#pd_deltavix = pd.concat([c_deltavixa, c_deltavixb, p_deltavixa, p_deltavixb], axis=1)
#pd_deltavix.index.name = "OFFATM"        
pd_deltavix
```

```python
import yfinance as yf

def obtener_precio_futuro_sp500():
    # Definir el ticker para el futuro del S&P 500 de expiraciÃ³n mÃ¡s cercana
    # El ticker se formatea como 'ES=F' en Yahoo Finance
    ticker_futuro = 'ES=F'
    
    # Obtener los datos del futuro usando yfinance
    datos_futuro = yf.Ticker(ticker_futuro)
    
    # Obtener el precio actual del futuro
    precio_futuro1 = datos_futuro.history(period='1m')['Close'].iloc[-1]
    print (precio_futuro1)
    #precio_futuro = datos_futuro.history(period='1m').Close.iloc[-1]
    #es_yf = yf.download(tickers="ES=F", interval="1m")
    es_yf = yf.download(tickers="SPY", interval="1m")
    precio_futuro = es_yf.iloc[-1]
    return precio_futuro

# Llamar a la funciÃ³n y mostrar el precio del futuro mÃ¡s cercano
precio = obtener_precio_futuro_sp500()
print(f"El precio del futuro del S&P 500 de expiraciÃ³n mÃ¡s cercana es: {precio}")
```

```python
import yfinance as yf
import datetime

def obtener_opciones_expiracion_mas_cercana(ticker='^SPX'):
#def obtener_opciones_expiracion_mas_cercana(ticker='ES=F'):
    # Obtener el objeto del Ã­ndice o futuro
    activo = yf.Ticker(ticker)
    
    # Obtener la lista de fechas de expiraciÃ³n disponibles
    expiraciones = activo.options
    
    # Obtener la fecha de expiraciÃ³n mÃ¡s cercana a hoy
    fecha_actual = datetime.datetime.now().date()
    fecha_expiracion_cercana = None
    
    for fecha in expiraciones:
        fecha_dt = datetime.datetime.strptime(fecha, '%Y-%m-%d').date()
        if fecha_dt >= fecha_actual:
            fecha_expiracion_cercana = fecha
            break

    if not fecha_expiracion_cercana:
        raise ValueError("No se encontraron opciones con expiraciÃ³n cercana.")

    # Obtener las opciones de la fecha de expiraciÃ³n mÃ¡s cercana
    opciones_cercanas = activo.option_chain(fecha_expiracion_cercana)

    # Extraer puts y calls
    calls = opciones_cercanas.calls
    puts = opciones_cercanas.puts

    return calls, puts

# Llamar a la funciÃ³n para obtener las opciones del S&P 500 (^SPX) de expiraciÃ³n mÃ¡s cercana
calls, puts = obtener_opciones_expiracion_mas_cercana()
print("Opciones Call mÃ¡s cercanas:\n", calls.head())
print("Opciones Put mÃ¡s cercanas:\n", puts.head())
```
