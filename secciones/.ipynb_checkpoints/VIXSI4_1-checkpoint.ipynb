### Estrategia VIXSI Intradía

La estrategia VIXSI intradía incluye la señal VIXSI a final de día más la señal VIXSI intradía, teniendo en cuenta la volatilidad intradía.

Se utiliza el dataset "intravixsi.csv", que contiene ambas señales y el SPY en intervalo de cinco minutos en el periodo 2022-2024.

Los datos para calcular la señal VIXSIintradia y SPY se han extraído del CBOE.

En este notebook se comparan ambas estrategias con SPY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from modulos.tfm_util import metrica_cartera
from modulos.tfm_util import drwdown

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
intravixsi = pd.read_csv("../datos/intravixsi.csv", index_col=0)
intravixsi.index = pd.to_datetime(intravixsi.index)
intravixsi["retspy"] = intravixsi.SPY.pct_change()

plt.figure(figsize=(10,6))
(1 + intravixsi.intraVIXSI.shift(1) * intravixsi.retspy).cumprod().plot(label="VIXSI_intradia")
(1 + intravixsi.VIXSI.shift(1) * intravixsi.retspy).cumprod().plot(label="VIXSI")
(1 + intravixsi.retspy).cumprod().plot(label="holding SPY")
plt.legend()
plt.plot()
plt.savefig('vixsi_intradia.png')
```

```python
intravixsi["estrategia"] = (1 + intravixsi.intraVIXSI.shift(1) * intravixsi.retspy).cumprod()
intravixsi["benchmark"] = (1 + intravixsi.retspy).cumprod()
df_vixsi1d = intravixsi.resample('D').last().dropna()

ret_ann, ret_ann_b, r_sharpe, r_sharpe_b = metrica_cartera(df_vixsi1d.estrategia, df_vixsi1d.benchmark)
maxdwn, *_ = drwdown (df_vixsi1d.estrategia)
maxdwn_b, *_ = drwdown (df_vixsi1d.SPY)

estrategias = ["VIXSI_intra", "HOLD"]
valores = [[ret_ann, r_sharpe, maxdwn], [ret_ann_b, r_sharpe_b, maxdwn_b]]
df_result = pd.DataFrame(valores, index= estrategias)
df_result.columns = ["RetAnual","RSharpe","MaxDwd"]

df_result = df_result.round(2)
df_result.to_clipboard()
df_result
```
