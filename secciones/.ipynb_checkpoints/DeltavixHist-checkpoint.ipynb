```python
# Se optimizan los precios con un delta operador para conseguir unos drawdowns objetivo
# y se generan los trades con dichos precios óptimos con el siguiente valor no visto siguiendo 
# con un walk forward de ventana expansiva con un único dato de test (el valor siguiente)


import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time
from pathlib import Path

def vix_investing ():
    '''
    
    '''
    
    url = "https://es.investing.com/indices/volatility-s-p-500-chart"
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
    else:
        print("no se ha podido cargar")
    soup = BeautifulSoup(html, "html.parser")
    vix_inv = soup.find('div', {'data-test': 'instrument-price-last'}).text
    fecha_inv = soup.find('time', {'data-test': 'trading-time-label'}).text
    vix_inv = float(vix_inv.replace(',', '.'))
    return vix_inv

def generar_vixsi_1m ():
    '''

    '''

    spy_yf = yf.download(tickers="SPY", interval="1m")
    ultima_fecha = spy_yf.index[-1]

    hora_exacta = ultima_fecha.time()

    df_vacio = True
    num_dias = 0
    while df_vacio:
      num_dias += 1
      if num_dias >= 15:
        print("no hay datos pasados de menos 15 días a hoy")
        return

      penultimo_dia_misma_hora = ultima_fecha - pd.Timedelta(days=num_dias)
      spy_ant = spy_yf.loc[
          (spy_yf.index.date == penultimo_dia_misma_hora.date()) & (spy_yf.index.time == hora_exacta)
      ]
      df_vacio = spy_ant.empty

    retspy = (spy_yf.iloc[-1].Close - spy_ant.iloc[-1].Close)/spy_ant.iloc[-1].Close

    vix_inv = vix_investing ()
    #fvixsi = np.abs(retspy) - vix_inv/(np.sqrt(256)*100)
    fvixsi = retspy - vix_inv/(np.sqrt(256)*100)
    vixsi = float(np.where ((fvixsi > 0), -1, 1))
    #20241005
    if retspy < 0:
        fvixsi = vix_inv/(np.sqrt(256)*100) - retspy

    #pd_vixsi1 = pd.DataFrame([vixsi], index=[ultima_fecha])
    #pd_vixsi1.columns = ["VIXSI"]
    #pd_vixsi = pd.concat([pd_vixsi, pd_vixsi1], axis=0)
    return (spy_yf.iloc[-1].Close, retspy, fvixsi)

def get_delta(datos, ret_activo, fvixsi):
    '''

    '''

    datos1 = datos.copy()
    datos1["xretsig"] = datos1.xrets.shift(-1)
    datos1.dropna(inplace=True)

    if ret_activo < 0:
        #if ret_activo > 0:
        #    datos1.NVIXSI = datos1.PVIXSI

        qnvixsi = datos1[datos1.xrets < 0].NVIXSI.rank(pct=True)
        NVIXSI = pd.concat(
                    [qnvixsi, datos1.loc[qnvixsi.index].NVIXSI,
                     datos1.loc[qnvixsi.index].xrets,
                     datos1.loc[qnvixsi.index].xretsig],
                     axis=1)
        NVIXSI.columns = ["QVIXSI","FVIXSI","xrets","xretsig"]
#20241109 revision
        if fvixsi >= NVIXSI.FVIXSI.min():
            qvixsi = NVIXSI[NVIXSI.FVIXSI<fvixsi].QVIXSI.max()
        else:
            qvixsi = 0
        
    else:
        qpvixsi = datos1[datos1.xrets > 0].PVIXSI.rank(pct=True)
        PVIXSI = pd.concat(
                    [qpvixsi, datos1.loc[qpvixsi.index].PVIXSI,
                     datos1.loc[qpvixsi.index].xrets,
                     datos1.loc[qpvixsi.index].xretsig],
                     axis=1)
        PVIXSI.columns = ["QVIXSI","FVIXSI","xrets","xretsig"]
#20241109 revision
        if fvixsi >= PVIXSI.FVIXSI.min():
            qvixsi = PVIXSI[PVIXSI.FVIXSI<fvixsi].QVIXSI.max()
        else:
            qvixsi = 0
#20241109 revision
    if ret_activo < 0:
        if qvixsi >= 0.5:
            # alcista
            qvixsi = (qvixsi - 0.5)*2
        else:
            # bajista
            qvixsi = (0.5 - qvixsi)*2 * -1
    else:
        if qvixsi >= 0.5:
            # bajista
            qvixsi = (qvixsi - 0.5)*2 * -1
        else:
            # alcista
            qvixsi = (0.5 - qvixsi)*2
    return qvixsi

#solo interesa Deltaindicators(Positivo,Negativo de 0 a 1, que indica la probabilidad de cambio)
def ini_deltavix (desde, hasta):
    '''
    
    '''

    fichero = f"datos2/deltavix-{desde}_{hasta}.csv"
    fichero = Path(fichero)

    if fichero.is_file():
        #datos = pd.read_csv(f"datos2/deltavix-{desde}_{hasta}.csv", index_col=0, parse_dates=True)
        datos = pd.read_csv(fichero, index_col=0, parse_dates=True)
        return datos

    spy_datos = yf.download("SPY", start=desde, end=hasta)
    vix_datos = yf.download("^VIX", start=desde, end=hasta)
    #spx_datos = yf.download("^SPX", start=desde, end=hasta)
    spx_datos = yf.download("SPY", start=desde, end=hasta)

    datos_all = pd.concat([spy_datos.Close, vix_datos.Close, spx_datos.Close], axis=1).dropna()
    datos_all.columns = ["SPY","VIX","SPX"]
    datos_all["rets"] = datos_all.SPY.pct_change()
    datos_all["xrets"] = datos_all.SPX.pct_change()
    datos_all["PVIXSI"] = datos_all.rets - datos_all.VIX/(np.sqrt(256)*100)
    datos_all["NVIXSI"] = datos_all.rets + datos_all.VIX/(np.sqrt(256)*100)
    #datos_all["FVIXSI"] = datos.rets - datos.VIX/(np.sqrt(256)*100)
    #datos_all["VIXSI"] = np.where ((datos.FVIXSI > 0), -1, 1)
    datos_all.dropna(inplace=True)
    #datos_all.to_csv(f"datos/deltavix-{desde}_{hasta}.csv")
    datos_all.to_csv(fichero)

    return datos_all

def get_deltavix (datos_all, desde_train, hasta_train):
    
    if desde_train == 0:
        fichero_train = f"datos2/deltavix_{hasta_train}_train.csv"
    else:
        fichero_train = f"datos2/deltavix_{desde_train}_{hasta_train}_train.csv"

    fichero_train = Path(fichero_train)
    if fichero_train.is_file():
        datos_train = pd.read_csv(fichero_train, index_col=0, parse_dates=True)
        return datos_train
    

    datos = datos_all.copy()
    datos = datos.iloc[desde_train:hasta_train]
    #datos["QVIXSI"] = datos.FVIXSI.rank(pct=True)
    datos.dropna(inplace=True)
    datos["qvixsi"] = -999
    
    ini = time.time()
    for i in range(len(datos)):
        ret_activo = datos.iloc[i]["xrets"]
        if ret_activo > 0:
            fvixsi = datos.iloc[i]["PVIXSI"]
        else:
            fvixsi = datos.iloc[i]["NVIXSI"]

        qvixsi = get_delta(datos, ret_activo, fvixsi)
        datos.iloc[i,datos.columns.get_loc("qvixsi")] = qvixsi

    datos.to_csv(fichero_train)
    return datos

def rsharpe (srent):
    rent = srent.mean() * 252
    var = srent.std() * np.sqrt(252)
    rsharpe = rent/var
    eq_rent = srent.cumsum()
    drwd = (eq_rent - eq_rent.cummax())
    maxdrwd = drwd.min()
    return (rsharpe, maxdrwd)
```

```python

desde = "2008-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")
hasta = "2024-11-13"
datos = ini_deltavix (desde, hasta)
num_dias_test = 1700
#VENTA DE PUT
ini = time.time()
ventaput = pd.DataFrame()
for num_dias in range(num_dias_test):
    hasta_train = len(datos) - (num_dias_test - num_dias + 1)
    desde_train = 0
    
    #datos train hasta_train (sin incluirlo)
    datos_train = get_deltavix (datos, desde_train, hasta_train)

    #qvixsi del último día de train (día actual en tiempo real)
    retact = datos_train.iloc[-1]["xrets"]
    if retact > 0:
        fvixsi = datos_train.iloc[-1]["PVIXSI"]
    else:
        fvixsi = datos_train.iloc[-1]["NVIXSI"]        
    qvixsi = get_delta(datos_train, retact, fvixsi)
    
    v_sl = 0.01
    v_maxdd = -0.04
    v_rshp = 1    
    for k in range(-35,40,5):
        #! para ventas los strikes ITM se ignoran (precio=9999)
        #if k > 0:
        #    continue
            
        for i in range(-10,10):
            if qvixsi < i/10 or qvixsi > (i + 1)/10:
                continue
            # inicio deltavix.fit
            minmax_delta = 500
            qrets = datos_train[(datos_train.qvixsi.shift(1)>i/10)&(datos_train.qvixsi.shift(1)<=(i+1)/10)][["SPX","xrets"]]
            #rmax = (-2.5 + k)/qrets.SPX[-1]
            #se coge el punto medio del rango 0.5 como valor medio para evaluar el delta strike 0 (k==0) como el primer
            #strike en precio (ATM) en put
            rmax = (k + 2)/datos_train.SPX[-1]

            #el % del strike con pérdidas no puede ser superior al % del stoploss
            if rmax <= -v_sl:
                break

            qrets["rets"] = 0
            qrets.loc[qrets.xrets<rmax,"rets"] = qrets[qrets.xrets<rmax].xrets - rmax
            qrets.loc[qrets.xrets<-v_sl,"rets"] = -v_sl - rmax

            for delta in range(500):
                precio = np.abs(qrets.rets.mean()) + qrets.rets.var() * delta
                qrets["rets_obj"] = qrets.rets + precio
                rshp, maxdw = rsharpe(qrets.rets_obj)
                if maxdw > v_maxdd and rshp > v_rshp:
                    minmax_dd = maxdw
                    minmax_i = i
                    minmax_j = i + 1
                    minmax_delta = delta
                    minmax_precio = precio
                    minmax_rs = rshp
                    break
            # fin deltavix.fit
            if minmax_delta < 500:
                retsig = datos.iloc[hasta_train]["xrets"]
                fechasig = datos.index[hasta_train]
                minmax_gain = precio                        
                if retsig < rmax:
                    minmax_gain = retsig - rmax + precio
                if retsig < -v_sl:
                    minmax_gain = -v_sl - rmax + precio

                deltaput = [num_dias, fechasig, k, minmax_i/10, minmax_j/10, minmax_delta, minmax_precio, minmax_rs, minmax_dd, minmax_gain]
                #print ("k:", k, "i:", minmax_i, "j:", minmax_i + 1, "delta:",minmax_delta,"rsharpe:",minmax_rs,"maxddown:",minmax_dd)
                ventaput = pd.concat([ventaput, pd.DataFrame([deltaput])], axis=0)
            #else:
            #    print ("k:", k, " no tiene solución", i, i+drango)
    #print("dia: ", num_dias, " tarda:", time.time() - ini)
if len(ventaput) > 0:
    ventaput.columns = ["num_dias", "fecha", "strike", "d_qvixsi", "h_qvixsi", "delta", "precio", "rsharpe","dwdown","gain"]
    ventaput.reset_index(inplace=True,drop=True)
    ventaput.to_csv("datos2/ventaput.csv")
```

```python
def rsharpe (srent):
    rent = srent.mean() * 252
    var = srent.std() * np.sqrt(252)
    rsharpe = rent/var
    eq_rent = srent.cumsum()
    drwd = (eq_rent - eq_rent.cummax())
    maxdrwd = drwd.min()
    return (rsharpe, maxdrwd)

desde = "2008-01-01"
#hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")
hasta = "2024-11-20"
#hasta = (df_op.index.unique().max()+ timedelta(1)).strftime("%Y-%m-%d")
#desde = df_op.index.unique().min().strftime("%Y-%m-%d")
datos = ini_deltavix (desde, hasta)
#num_dias_test = len(df_op.index.unique())
num_dias_test = 1700
#print (datos.index[-1])
#COMPRA DE CALL
compracall = pd.DataFrame()
ini = time.time()
for num_dias in range(num_dias_test):
    hasta_train = len(datos) - (num_dias_test - num_dias)
    desde_train = 0
    #print ("desde-hasta train:",desde_train,datos.iloc[hasta_train])
    datos_train = get_deltavix (datos, desde_train, hasta_train)

    v_maxdd = -0.04
    v_rshp = 1
    
    retact = datos_train.iloc[-1]["xrets"]
    if retact > 0:
        fvixsi = datos_train.iloc[-1]["PVIXSI"]
    else:
        fvixsi = datos_train.iloc[-1]["NVIXSI"]        
    qvixsi = get_delta(datos_train, retact, fvixsi)

    for k in range(-35,40,5):
        for i in range(-10,10):
            if qvixsi < i/10 or qvixsi > (i+1)/10:
                continue
            minmax_delta = 500
            qrets = datos_train[(datos_train.qvixsi.shift(1)>i/10)&(datos_train.qvixsi.shift(1)<=(i+1)/10)][["SPX","xrets"]]
            rmax = (-2 + k)/qrets.SPX[-1]

            qrets["rets"] = 0
            qrets.loc[qrets.xrets>=rmax,"rets"] = qrets[qrets.xrets>=rmax].xrets - rmax
            qrets.loc[qrets.xrets<rmax,"rets"] = 0

            for delta in range(500):
                precio = np.abs(qrets.rets.mean()) - qrets.rets.var() * delta
                qrets["rets_obj"] = qrets.rets - precio
                rshp, maxdw = rsharpe(qrets.rets_obj)
                if maxdw > v_maxdd and rshp > v_rshp:
                    minmax_dd = maxdw
                    minmax_i = i
                    minmax_j = i + 1
                    minmax_delta = delta
                    minmax_precio = precio
                    minmax_rs = rshp
                    break

            if minmax_delta < 500:
                #print (datos_train.index[-1])
                retsig = datos.iloc[hasta_train]["xrets"]
                fechasig = datos.index[hasta_train]
                minmax_gain = - precio
                if retsig >= rmax:
                    minmax_gain = retsig - rmax - precio

                deltacall = [num_dias, fechasig, k, minmax_i/10, minmax_j/10, minmax_delta, minmax_precio, minmax_rs, minmax_dd, minmax_gain]
                #print ("k:", k, "i:", minmax_i, "j:", minmax_i + 1, "delta:",minmax_delta,"rsharpe:",minmax_rs,"maxddown:",minmax_dd)
                compracall = pd.concat([compracall, pd.DataFrame([deltacall])], axis=0)

            #else:
            #    print ("k:", k, " no tiene solución", i, i+1)
                    
    #print("dia: ", num_dias, " tarda:", time.time() - ini)

compracall.columns = ["num_dias", "fecha", "strike", "d_qvixsi", "h_qvixsi", "delta", "precio", "rsharpe","dwdown","gain"]
compracall.reset_index(inplace=True,drop=True)
compracall.to_csv("datos2/compracall.csv")
```

```python
ventaput = pd.read_csv("datos2/ventaput.csv", index_col=0, parse_dates=True)
ventacall = pd.read_csv("datos2/ventacall.csv", index_col=0, parse_dates=True)
compraput = pd.read_csv("datos2/compraput.csv", index_col=0, parse_dates=True)
compracall = pd.read_csv("datos2/compracall.csv", index_col=0, parse_dates=True)

df_op = pd.read_csv("opciones/SPY2024-11.csv", index_col=0, parse_dates=True)
df_op["fecha"] = pd.to_datetime(df_op["quote_date"])
df_op["fexp"] = pd.to_datetime(df_op["expiration"])
df_op["indice"] = df_op["fexp"]
df_op = df_op[(df_op["quote_date"]!=df_op["expiration"])]
df_op['delta'] = (df_op['fexp'] - df_op['fecha']).dt.days
df_op.reset_index(inplace=True, drop=True)
df_op.set_index('indice', inplace=True)
df_op = df_op.loc[df_op.groupby('fecha')['delta'].idxmin()]
df_op["CATM"] = df_op['underlying_bid_eod'] - df_op['underlying_bid_eod']%5
df_op["PATM"] = df_op['underlying_bid_eod'] + (5 - df_op['underlying_bid_eod']%5)
df_op["CSTRIKE"] = df_op.strike - df_op.CATM
df_op["PSTRIKE"] = df_op.strike - df_op.PATM
df_op = df_op[np.abs(df_op.CSTRIKE)<=50]
df_op = df_op[["fecha","fexp","underlying_ask_eod",
                    "option_type","strike","bid_eod","ask_eod","CSTRIKE","PSTRIKE"]]
df_op.columns=["FECHA","FEXP","ACTIVO","TIPO","STRIKE","BID","ASK","CSTRIKE","PSTRIKE"]
df_op["BID"]=df_op["BID"]/df_op["ACTIVO"]
df_op["ASK"]=df_op["ASK"]/df_op["ACTIVO"]
```

```python
#CALL
#compra
df_op1 = df_op.loc[(df_op.CSTRIKE == 0) & (df_op.TIPO=="C")]
df_op1['delta'] = (df_op1['FEXP'] - df_op1['FECHA']).dt.days
df_op1.reset_index(inplace=True, drop=True)
df_op1 = df_op1.loc[df_op1.groupby('FECHA')['delta'].idxmin()]
df_op1.set_index('FEXP', inplace=True)
df_op1
fechas_test = df_op.index.unique().to_list()
compracall1 = compracall[compracall.strike==0]
compracall1["fechasig"] = pd.to_datetime(compracall1["fecha"])
compracall1.set_index('fechasig', inplace=True)
compracall1 = compracall1.loc[fechas_test]
tradescc = pd.concat([df_op1, compracall1], axis=1)
tradescc_ok = tradescc[tradescc.ASK <= tradescc.precio]
tradescc_ok["gain"] = tradescc_ok["gain"] + tradescc_ok.precio - tradescc_ok.ASK
tradescc_ok.gain.sum(), len(tradescc_ok), len(fechas_test)
```
