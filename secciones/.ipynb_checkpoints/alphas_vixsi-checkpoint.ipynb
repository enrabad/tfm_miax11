## Analizar alpha generado por la señal VIXSI en el activo SPY

Una forma de validar la señal es ver cuanto alpha genera la estrategia que utiliza la señal para invertir diariamente en largo (vixsi=1) o en corto (vixsi=-1) en un activo respecto a mantenerse invertido en el propio activo.

Incialmente se utiliza como activo el ETF que replica al SP500 (SPY).

la fórmula del alpha será simplemente el exceso de retorno de la señal respecto al activo:

**alpha_vixsignal = rentabilidad vixsignal - rentabilidad activo**

El alpha se agrupa por año, y se evaluará la media y el ratio de sharpe del alpha anual.

Los datos del SPY se recogen de yfinance y se ajustan a los dividendos.


```python
#%pip install --upgrade investpy

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
```

### alpha_vixsi_spy

Una forma de validar la señal es ver cuanto alpha genera la estrategia que utiliza la señal para invertir en largo (vixsi=1) o en corto (vixsi=-1) un activo respecto a mantenerse invertido en el activo.

la fórmula del alpha se basará en el alpha de Jensen (asignando 0 a la rentabilidad libre de riesgo):

rentabilidad señal - Beta_señal_activo * rentabilidad activo

El alpha se agrupa por año, y se evalua la media .

Los datos del SPY se recogerán de yfinance y se ajustarán a los dividendos.


```python
def f_alpha_act(datos, estrategia="estrategia", benchmark="SPY"):
    '''

    '''

    srets = np.log(datos[estrategia]).diff().fillna(0)
    brets = np.log(datos[benchmark]).diff().fillna(0)

    #w_brets = brets.loc[srets.index]
    cov_rets = np.cov(srets.T, brets)[0,1:]
    var_brets = brets.var()
    beta = cov_rets / var_brets
    rent_srets = srets.sum()
    rent_brets = brets.sum()
    alpha = rent_srets - beta * rent_brets    
    return (alpha)

def f_alpha(datos, estrategia="estrategia", benchmark="SPY"):
    '''

    '''

    srets = np.log(datos[estrategia]).diff().fillna(0)
    brets = np.log(datos[benchmark]).diff().fillna(0)

    rent_srets = srets.sum()
    rent_brets = brets.sum()
    
    alpha = rent_srets - rent_brets    
    return (alpha, rent_brets, rent_srets)

#activo = "KLAC"
activo1 = "MSFT"
activo2 = "SPY"
activo1 = "KLAC"
activo2 = "SPY"
benchmark = activo1

desde = "2008-01-01"
hasta = "2024-01-01"

#Los datos históricos del VIXSI se extraen de un fichero
df_vixsi = pd.read_csv("../datos/VIXSIyf.csv", index_col=0)
df_vixsi.index =  pd.to_datetime(df_vixsi.index)

desde = df_vixsi.index[0].strftime("%Y-%m-%d")
hasta = df_vixsi.index[-1].strftime("%Y-%m-%d")

#Los datos del SPY se recogerán de yfinance y se ajustarán a los dividendos
spy_datos1 = yf.download(activo1, start=desde, end=hasta, auto_adjust=True)
spy_datos2 = yf.download(activo2, start=desde, end=hasta, auto_adjust=True)

datos = pd.concat([spy_datos1.Close, spy_datos2.Close, df_vixsi.VIXSI], axis=1).dropna()
datos.columns = [activo1,activo2,"vixsi"]

#Se calculan los retornos acumulados de las estrategias holding y vixsi
datos["estrHolding"] = (1 + datos[activo1].pct_change()).cumprod()
datos["estrvixsi"] = (1 + datos.vixsi.shift(1) * datos[activo1].pct_change()).cumprod()
datos["Estralpha"] = (1 + (datos.vixsi.shift(1) * datos[activo1].pct_change() - datos[activo2].pct_change())/2).cumprod()


#Se generan los alphas por año
años = sorted(datos.index.year.unique())
alpha = np.zeros(len(años))
señal = np.zeros(len(años))
activo = np.zeros(len(años))
for i in range(len(años)-2):
    #alpha[i], activo[i], señal[i] = f_alpha(datos.loc[str(años[i])],estrategia="estrvixsi",benchmark=benchmark)
    alpha[i], activo[i], señal[i] = f_alpha(datos.loc[str(años[i]):str(años[i]+2)],estrategia="estrvixsi",benchmark=benchmark)
    #alpha[i], activo[i], señal[i] = f_alpha(datos.loc[str(años[i])],estrategia=activo1,benchmark=benchmark)
    
alpha = pd.DataFrame(alpha, index=años)
activo = pd.DataFrame(activo, index=años)
señal = pd.DataFrame(señal, index=años)

print("media anual del alpha negativo:", alpha[alpha<0].mean().values)

plt.figure(1)
alpha.plot()
plt.title('Alpha de la estrategia VIXSI respecto al SPY')
plt.axhline(y=0, color='b', linestyle='--')
plt.show()

plt.figure(2)
estrategias = datos.filter(regex='^estr')
#estrategias.loc["2012":"2016"].plot()
estrategias.plot()
plt.title('Equity estrategias')
plt.show()
```

```python
def f_alpharol(s, b, isalpha=0):
    w_b = b.loc[s.index]
    srets = np.log(s).diff().fillna(0)
    brets = np.log(w_b).diff().fillna(0)
    #rent_srets = srets.sum()/len(srets)
    #rent_brets = w_brets.sum()/len(w_brets)
    rent_srets = srets.sum()
    rent_brets = brets.sum()
    alpha = rent_srets -  rent_brets
    if isalpha==0:
        return (alpha)
    elif isalpha==1:
        return (rent_brets)
    elif isalpha==2:
        return (rent_srets)

alpharol = datos.estrvixsi.rolling(5).apply(lambda x: f_alpharol(x,datos.benchmark))
benchrol = datos.estrvixsi.rolling(5).apply(lambda x: f_alpharol(x,datos.benchmark,isalpha=1))
estrrol = datos.estrvixsi.rolling(5).apply(lambda x: f_alpharol(x,datos.benchmark,isalpha=2))

alpharol.corr(alpharol.shift(-5))
alpharolcob = alpharol[::5]
benchrolcob = benchrol[::5]
estrrolcob = estrrol[::5]
((alpharolcob > 0)*alpharolcob.shift(-1) + (alpharolcob < 0)*benchrolcob.shift(-1)).cumsum().plot()
benchrolcob.shift(-1).cumsum().plot()
estrrolcob.shift(-1).cumsum().plot()
```

```python
datos["brets"] = np.log(datos.estrHolding).diff()
datos["srets"] = np.log(datos.estrvixsi).diff()
datos["alpha"] = datos["srets"] - datos["brets"]
datos.dropna(inplace=True)
```

```python
alphai = datos.alpha.rolling(5).sum()[::5]
bretsi = datos.brets.rolling(5).sum()[::5]
sretsi = datos.srets.rolling(5).sum()[::5]
((alphai >= 0)*alphai.shift(-1) + (alphai < 0)*bretsi.shift(-1)).cumsum().plot()
((alphai >= 0)*sretsi.shift(-1) + (alphai < 0)*bretsi.shift(-1)).cumsum().plot()
sretsi.cumsum().plot()
```
