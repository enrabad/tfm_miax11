### Estrategia VIXSI Intradía

La estrategia VIXSI intradía es similar a la estrategia VIXSI (que se calcula a final de día), solo que el cálculo de la señal -1 se hace en tiempo de mercado cada minuto, con lo que se tiene en cuenta la volatilidad intradía para desactivar la señal (volverla a 1).

El histórico se puede acceder desde una una api.
La fecha de inicio es del 11-12-2024, y se puede acceder a los datos actualizados a través de la api expuesta en este notebook,
que también compara la equity de la estrategia con la de estar en holding en SPY.

### Histórico intradía VIXSI ajustado a la volatilidad diaria del VIX + SPY y estrategia VIXSI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests


base_url = 'http://5.250.185.6:5001/api/datos'
headers = {'Authorization': "7qB9mZ3p"}
response = requests.get(base_url, headers=headers)
if response.status_code == 200:
        data = response.json()
        df_data = pd.DataFrame(data)
        df_data.set_index("index", inplace=True)
else:
        df_data = None
        print('Error al hacer la solicitud:', response.status_code, response.text)

if df_data is None:
    print("no hay datos")
else:
    df_data.index = pd.to_datetime(df_data.index)
    df_data = df_data.sort_index()
    df_vixsi = df_data

    plt.figure(figsize=(10,6))
    (1 + df_vixsi.VIXSI.shift(1) * df_vixsi.SPY.pct_change()).cumprod().plot(label="VIXSI_intradia")
    (1 + df_vixsi.SPY.pct_change()).cumprod().plot(label="holding SPY")
    plt.legend()
    plt.plot()
```
