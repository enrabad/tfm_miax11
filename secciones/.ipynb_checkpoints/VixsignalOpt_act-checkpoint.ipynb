```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time

desde = "2008-01-01"
hasta = "2024-10-01"

#df_vixsi = pd.read_csv("/content/drive/MyDrive/vixsi/VIXSIyf.csv", index_col=0)
#df_vixsi.index =  pd.to_datetime(df_vixsi.index)
#desde = df_vixsi.index[0].strftime("%Y-%m-%d")
#hasta = df_vixsi.index[-1].strftime("%Y-%m-%d")

spy_datos = yf.download("SPY", start=desde, end=hasta)
vix_datos = yf.download("^VIX", start=desde, end=hasta)
spx_datos = yf.download("^SPX", start=desde, end=hasta)

datos = pd.concat([spy_datos.Close, vix_datos.Close, spx_datos.Close], axis=1).dropna()
datos.columns = ["SPY","VIX","SPX"]
datos["rets"] = datos.SPY.pct_change()
datos["xrets"] = datos.SPX.pct_change()
datos["FVIXSI"] = datos.rets - datos.VIX/(np.sqrt(256)*100)
datos["VIXSI"] = np.where ((datos.FVIXSI > 0), -1, 1)
datos["QVIXSI"] = datos.FVIXSI.rank(pct=True)
datos.dropna(inplace=True)
datos1 = datos.copy()
datos1["xretsig"] = datos1.xrets.shift(-1)
datos1.dropna(inplace=True)

activo = datos.iloc[-1].SPX
ret_activo = datos.iloc[-1].xrets
strike = activo*(1 + ret_activo)
#porce = (strike/activo - 1)/ret_activo
print(f"SPX: {datos.iloc[-1].SPX}")
print(f"%SPX: {datos.iloc[-1].xrets:.2f}")
#print(f"%STRIKE: {porce:.2f}")
#if porce < 0:
#    porce = 1

xrets = datos.iloc[-1].SPX
qvixsi = datos.iloc[-1].QVIXSI
#df qvixsi < 0.60:
datos2 = datos1[datos1.QVIXSI<qvixsi]
print(f"strikeATM-: {datos.iloc[-1].SPY*datos.iloc[-1].xrets:.2f}")
print("venta PUT probabilidad retorno siguiente inferior al strikeATM-:",
      len(datos2[datos2.xretsig < datos2.xrets])/len(datos2))
print("venta PUT probabilidad retorno siguiente inferior a mitad del decrementoATM-/2:",
      len(datos2[datos2.xretsig < (datos2.xrets/2)])/len(datos2))
#else:
datos2 = datos1[datos1.QVIXSI>qvixsi]
print(f"strikeATM+: {datos.iloc[-1].SPY*datos.iloc[-1].xrets:.2f}")
print("venta CALL probabilidad retorno siguiente superior al strikeATM+:",
      len(datos2[datos2.xretsig > datos2.xrets])/len(datos2))
print("venta CALL probabilidad retorno siguiente superior a mitad del decrementoATM+/2:",
      len(datos2[datos2.xretsig > (datos2.xrets/2)])/len(datos2))
```
