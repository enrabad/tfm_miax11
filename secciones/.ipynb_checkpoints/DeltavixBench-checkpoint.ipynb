```python
#El objetivo de este indicador es ayudar a crear estrategias credit y debit spreads tanto para call como para put


import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time
from pathlib import Path
```

```python
df_op202401 = pd.read_csv("opciones/SPY2024-01.csv", index_col=0, parse_dates=True)
df_op202402 = pd.read_csv("opciones/SPY2024-02.csv", index_col=0, parse_dates=True)
df_op202403 = pd.read_csv("opciones/SPY2024-03.csv", index_col=0, parse_dates=True)
df_op202404 = pd.read_csv("opciones/SPY2024-04.csv", index_col=0, parse_dates=True)
df_op202405 = pd.read_csv("opciones/SPY2024-05.csv", index_col=0, parse_dates=True)
df_op202406 = pd.read_csv("opciones/SPY2024-06.csv", index_col=0, parse_dates=True)
df_op202407 = pd.read_csv("opciones/SPY2024-07.csv", index_col=0, parse_dates=True)
df_op202408 = pd.read_csv("opciones/SPY2024-08.csv", index_col=0, parse_dates=True)
df_op202409 = pd.read_csv("opciones/SPY2024-09.csv", index_col=0, parse_dates=True)
df_op202410 = pd.read_csv("opciones/SPY2024-10.csv", index_col=0, parse_dates=True)
df_op202411 = pd.read_csv("opciones/SPY2024-11.csv", index_col=0, parse_dates=True)
```

```python
df_op = pd.concat([df_op202401,
                  df_op202402,
                  df_op202403,
                  df_op202404,
                  df_op202405,
                  df_op202406,
                  df_op202407,
                  df_op202408,
                  df_op202409,
                  df_op202410,
                  df_op202411,                   
                  ], axis=0)
```

```python
ventaput = pd.read_csv("datos2/ventaput.csv", index_col=0, parse_dates=True)
ventacall = pd.read_csv("datos2/ventacall.csv", index_col=0, parse_dates=True)
compraput = pd.read_csv("datos2/compraput.csv", index_col=0, parse_dates=True)
compracall = pd.read_csv("datos2/compracall.csv", index_col=0, parse_dates=True)

df_op["fecha"] = pd.to_datetime(df_op["quote_date"])
df_op["fexp"] = pd.to_datetime(df_op["expiration"])
df_op["indice"] = df_op["fexp"]
df_op = df_op[(df_op["quote_date"]!=df_op["expiration"])]
df_op['delta'] = (df_op['fexp'] - df_op['fecha']).dt.days
df_op.reset_index(inplace=True, drop=True)
df_op.set_index('indice', inplace=True)
df_op = df_op.loc[df_op.groupby('fecha')['delta'].idxmin()]
df_op["CATM"] = df_op['underlying_bid_eod'] - df_op['underlying_bid_eod']%5
df_op["PATM"] = df_op['underlying_bid_eod'] + (5 - df_op['underlying_bid_eod']%5)
df_op["CSTRIKE"] = df_op.strike - df_op.CATM
df_op["PSTRIKE"] = df_op.strike - df_op.PATM
df_op = df_op[np.abs(df_op.CSTRIKE)<=50]
df_op = df_op[["fecha","fexp","underlying_ask_eod",
                    "option_type","strike","bid_eod","ask_eod","CSTRIKE","PSTRIKE"]]
df_op.columns=["FECHA","FEXP","ACTIVO","TIPO","STRIKE","BID","ASK","CSTRIKE","PSTRIKE"]
#df_op["BID"]=df_op["BID"]/df_op["ACTIVO"]
#df_op["ASK"]=df_op["ASK"]/df_op["ACTIVO"]
```

```python
df_op1 = df_op.loc[(df_op.CSTRIKE == 0) & (df_op.TIPO=="C")]
df_op1['delta'] = (df_op1['FEXP'] - df_op1['FECHA']).dt.days
df_op1.reset_index(inplace=True, drop=True)
df_op1 = df_op1.loc[df_op1.groupby('FECHA')['delta'].idxmin()]
df_op1.set_index('FEXP', inplace=True)
df_op1["sigpuntos"] = df_op1["ACTIVO"].shift(-1) - df_op1["STRIKE"]
df_op1["ccgain"] = np.where(df_op1.sigpuntos > 0, df_op1.sigpuntos - df_op1.ASK, - df_op1.ASK) 
df_op1["vcgain"] = df_op1.ASK
df_op1["vcgain"] = np.where(df_op1.sigpuntos > 0, df_op1.ASK - df_op1.sigpuntos, df_op1.ASK)   
df_op1.ccgain.cumsum().plot()
df_op1.vcgain.cumsum().plot()
```

```python
df_op1["cASK"]=df_op1["ASK"]/df_op1["ACTIVO"]
fechas_test = df_op.index.unique().to_list()
compracall1 = compracall[compracall.strike==0]
compracall1["fechasig"] = pd.to_datetime(compracall1["fecha"])
compracall1.set_index('fechasig', inplace=True)
compracall1 = compracall1.loc[fechas_test]
tradescc = pd.concat([df_op1, compracall1], axis=1)
tradesvc_ok = tradescc[np.abs(tradescc.d_qvixsi) > 0.7]
tradesvc_ok.vcgain.cumsum().plot()
tradescc_ok = tradescc[np.abs(tradescc.d_qvixsi) < 0.3]
tradescc_ok.ccgain.cumsum().plot()
```

```python
df_op1 = df_op.loc[(df_op.CSTRIKE == 0) & (df_op.TIPO=="P")]
df_op1['delta'] = (df_op1['FEXP'] - df_op1['FECHA']).dt.days
df_op1.reset_index(inplace=True, drop=True)
df_op1 = df_op1.loc[df_op1.groupby('FECHA')['delta'].idxmin()]
df_op1.set_index('FEXP', inplace=True)
df_op1["sigpuntos"] = df_op1["ACTIVO"].shift(-1) - df_op1["STRIKE"]
df_op1["cpgain"] = np.where(df_op1.sigpuntos < 0, - df_op1.sigpuntos - df_op1.BID, - df_op1.BID) 
df_op1["vpgain"] = df_op1.BID
df_op1["vpgain"] = np.where(df_op1.sigpuntos < 0, df_op1.BID + df_op1.sigpuntos, df_op1.BID)   
df_op1.cpgain.cumsum().plot()
df_op1.vpgain.cumsum().plot()
```

```python
fechas_test = df_op.index.unique().to_list()
compraput1 = compraput[compraput.strike==0]
compraput1["fechasig"] = pd.to_datetime(compraput1["fecha"])
compraput1.set_index('fechasig', inplace=True)
compraput1 = compraput1.loc[fechas_test]
tradescp = pd.concat([df_op1, compraput1], axis=1)
tradesvp_ok = tradescp[np.abs(tradescp.d_qvixsi) < 0.3]
tradesvp_ok.vpgain.cumsum().plot()
tradescp_ok = tradescp[np.abs(tradescp.d_qvixsi) > 0.7]
tradescp_ok.cpgain.cumsum().plot()
```

```python
tradescc.loc[np.abs(tradescc.d_qvixsi) >= 0.3,"ccgain"] = 0
tradescc.loc[np.abs(tradescc.d_qvixsi) <= 0.7,"vcgain"] = 0
(tradescc.vcgain + tradescc.ccgain).cumsum().plot()
tradescp.loc[np.abs(tradescp.d_qvixsi) >= 0.3,"vpgain"] = 0
tradescp.loc[np.abs(tradescp.d_qvixsi) <= 0.7,"cpgain"] = 0
(tradescp.vpgain + tradescp.cpgain).cumsum().plot()
```
