# ANALISIS VIXSI

En este notebook se analiza una sencilla estrategia con el indicador VIXSI, observando los alphas generados en diferentes periodos.


```python
#Importamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta

%matplotlib inline
```

```python
desde = "2008-01-01"
hasta = datetime.now().strftime("%Y-%m-%d")
VIX = yf.download(tickers="^VIX", start=desde, end=hasta)
SPY = yf.download(tickers="SPY", start=desde, end=hasta, auto_adjust=False)
datos = pd.concat([VIX.Close,SPY.Close], axis=1).dropna()
datos.columns = ["VIX","SPY"]

def vixsignal (datos, vix="VIX", benchmark="SPY"):
    rets = datos[benchmark].pct_change()
    fvixsi = rets - datos[vix]/(np.sqrt(256)*100)
    #vixsi = datos[vix]/(np.sqrt(256)*100)
    vixsi = np.where ((fvixsi > 0), -1, 1)
    fvixsi1 = np.abs(rets) - datos[vix]/(np.sqrt(256)*100)
    vixsi1 = np.where (((rets < 0) & (fvixsi1>0)), 0, vixsi)
    return vixsi, vixsi1

vixsi_df = pd.DataFrame(index=datos.index)
vixsi, vixsi1 =  vixsignal (datos)
vixsi_df["SPY"] = datos.SPY
vixsi_df["VIXSI"] = vixsi
vixsi_df.dropna(inplace=True)
#vixsi_df.to_csv("../datos/VIXSIyf.csv")
vixsi_df["VIXSI1"] = vixsi1
```

### Cargar datos

```python
def cargar_datos ():
    '''
    Carga el indicador VIXSI desde ../datos/VIXSI.csv, SPY desde 
    
    '''

    vixsi_datos = pd.read_csv("../datos/VIXSI.csv", index_col=0)
    vixsi_datos.index = pd.to_datetime(vixsi_datos.index)
    desde = vixsi_datos.index[0].strftime("%Y-%m-%d")
    hasta = vixsi_datos.index[-1].strftime("%Y-%m-%d")
    spy_datos = yf.download("SPY", start=desde, end=hasta)

    datos = pd.concat([vixsi_datos.VIXSI, spy_datos.Close], axis=1).dropna()
    datos.columns = ["VIXSI","SPY"]

    datos.dropna(inplace=True)

    return datos

datos = cargar_datos()
```

### Construcción de la señal de trading
#### VIXSI es un indicador que refleja optimismo en el mercado cuando su valor es positivo.
Debido a que el indicador está inversamente correlacionado con los retornos del día siguiente, se implementa una señal de sentimiento contrario, donde se le da "-1" a los valores positivos del indicador y "1" a los valores negativos, poniéndose en corto el activo cuando el indicador es positivo.

```python
def vixsignal (datos):
    return np.where (datos.VIXSI > 0, -1, 1)

datos["vixsi"] = vixsignal(datos)
datos["estrategia"] = (1 + datos.vixsi * datos.SPY.pct_change().shift(-1)).cumprod()
datos["benchmark"] = (1 + datos.SPY.pct_change().shift(-1)).cumprod()
datos.dropna(inplace=True)
```

### Generación de alpha

Una forma de validar la señal es ver cuanto alpha genera la estrategia que utiliza la señal para invertir en largo (vixsi=1) o en corto (vixsi=-1) un activo respecto a mantenerse invertido en el activo.

la fórmula del alpha será la diferencia de retornos logaritmicos entre la estrategia y benchmark en un periodo determinado, rolando los datos para proporcionar una visión global del alpha en todo el histórico.
 

```python
def f_alpharol(s, b):
    w_b = b.loc[s.index]
    srets = np.log(s).diff().fillna(0)
    brets = np.log(w_b).diff().fillna(0)
    rent_srets = srets.sum()
    rent_brets = brets.sum()
    return rent_srets -  rent_brets

datos["estrategia"] = (1 + datos.vixsi * datos.SPY.pct_change().shift(-1)).cumprod()
datos["benchmark"] = datos.SPY
datos["alphas"] = datos.estrategia.rolling(252).apply(lambda x: f_alpharol(x,datos.benchmark))
```

Como se puede observar en los datos estadísticos, en promedio la estrategia genera un exceso de rentabilidad de un 9%, con un mínimo de -16% y un máximo de 72%, siendo el decil 0.5 del 5%

```python
periodos = [252, 252*2, 252*3]
palphas = []
nalpha = []
nspy = []
nestr = []

for periodo in periodos:
    datos["alphas"] = datos.estrategia.rolling(periodo).apply(lambda x: f_alpharol(x,datos.benchmark))
    datos["retSPY"] = np.log(datos.SPY).diff()
    datos["rretSPY"] = datos.retSPY.rolling(periodo).sum()
    datos["retestr"] = np.log(datos.estrategia).diff()
    datos["rretestr"] = datos.retestr.rolling(periodo).sum()
    palphas.append((datos.alphas > 0).sum()/len(datos))
    nalpha.append(((datos.alphas < 0) & (datos.rretSPY < 0)).sum()/len(datos))
    nspy.append((datos.rretSPY < 0).sum()/len(datos))
    nestr.append((datos.rretestr < 0).sum()/len(datos))
```

#### Se crea una función (built_model_...) para cada modelo y una función general (fit_eval_model) para evaluar todos los modelos

```python

def build_model_regresion(X):
    '''
    Se implementa el modelo de regresión lineal, con una sola capa densa de una neurona sin función de activación
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "REGRESION"
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(XX_tr.shape[1],XX_tr.shape[2])))
    model.add(BatchNormalization())
    model.add(keras.layers.Dense(1, activation="sigmoid"))
    
    return model, model_name

def build_model_LSTM(X):
    '''
    Se implementa el modelo LSTM con una sola capa LSTM de tamaño temporal XX.shape[1] y canales XX.shape[2]
    La salida es una capa densa de una sola neurona sin función de activación
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''

    model_name = "LSTM"
    model = keras.Sequential()

    #return_sequences=True. Devuelve como salida toda la secuencia temporal en lugar de la última para la entrada de la siguiente capa
    #Como salida devuelve la misma dimensión temporal que la entrada

    #model.add(keras.layers.LSTM(X.shape[1], input_shape=(X.shape[1],X.shape[2]), return_sequences=True))
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.LSTM(X.shape[1]*2, activation="relu", return_sequences=True))
    model.add(keras.layers.Flatten())
    model.add(Dense(X.shape[1], activation="relu"))
    model.add(Dense(1, activation="sigmoid"))

    return model, model_name

def build_model_GRU(X):
    '''
    Se implementa el modelo GRU de la misma manera que el modelo LSTM
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "GRU"
    model = keras.Sequential()

    #return_sequences=True. Devuelve como salida toda la secuencia temporal en lugar de la última para la entrada de la siguiente capa
    #Como salida devuelve la misma dimensión temporal que la entrada
    #model.add(keras.layers.GRU(X.shape[1], input_shape=(X.shape[1],X.shape[2],), return_sequences=True, activation="tanh"))
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.GRU(X.shape[1]*2, activation="relu", return_sequences=True))
    model.add(keras.layers.Flatten())
    model.add(Dense(X.shape[1], activation="relu"))
    model.add(Dense(1, activation="sigmoid"))

    return model, model_name

def build_model_CONV(X):
    '''
    Se implementa un modelo convolucional de dos capas de 16 y 8 filtros de tamaño 3 con pooling de 2 en cada capa para reducir la dimensión de las salidas
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''

    model_name = "CONV"
 
    model = keras.Sequential()
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    #model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='tanh', input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='relu', input_shape=(X.shape[1],X.shape[2])))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Convolution1D(filters=8, kernel_size=3, activation="relu"))
    model.add(keras.layers.Flatten())
    model.add(Dense(8, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
              
    return model, model_name

def build_model_CONVLSTM(X):
    '''
    Se implementa un modelo convolucional y recurrente (lstm), donde la primera capa será un modelo convolicional que como salida genere la misma
    dimensión que la entrada para que pase por un modelo convolucional 1D similar al modelo anterior
    
    Se le pasa la variable de entrada X para recoger las dimensiones temporal y activos
    
    '''
    
    model_name = "CONVLSTM"

    model = keras.Sequential()
    
    #Le añade return_sequences=True para pasar toda la secuencia temporal al la capa convolucional
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    #model.add(keras.layers.LSTM(X.shape[1], input_shape=(X.shape[1],X.shape[2],), return_sequences=True))
    model.add(keras.layers.LSTM(X.shape[1], return_sequences=True, activation="relu"))
    model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='relu', input_shape=(X.shape[1],X.shape[2])))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Convolution1D(filters=8, kernel_size=3, activation="relu"))
    model.add(keras.layers.Flatten())
    model.add(Dense(8, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    
    return model, model_name

def build_model_attention (X):
    '''
    Se implementa un modelo de atención con tres capas densas (query, value y key) para transformar la salida de la capa LSTM en vectores de consulta,
    de valor y de clave como entrada en la capa attention
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "ATTENTION"

    #XX = Input((X.shape[1],X.shape[2],))
    inputs = Input(shape=(X.shape[1],X.shape[2]))
    XX = BatchNormalization()(inputs)
    out = LSTM(10, return_sequences = True, activation="relu")(XX)

    query = Dense(X.shape[1], activation="relu")(out)
    value = Dense(X.shape[1], activation="relu")(out)
    key = Dense(X.shape[1], activation="relu")(out)

    out = Attention(use_scale=False)([query,value,key])
    out = Flatten()(out)
    out = Dense(10, activation="relu")(out)
    out = Dense(1, activation="sigmoid")(out)

    model = Model(inputs=XX, outputs=out)
    
    return model, model_name

def build_model_bidireccional (X):
    '''
    Se implementa un modelo bidireccional asimétrico (GRU para forward yt LSTM para backward)
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "BIDIRECCIONAL"

    model = keras.Sequential()

    forward_layer = keras.layers.GRU(20,activation='relu',return_sequences=True)
    backward_layer = keras.layers.LSTM(20, activation='relu',return_sequences=True, go_backwards=True)

    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.Bidirectional(forward_layer, backward_layer=backward_layer,
                                         #input_shape=(X.shape[1],X.shape[2])
             ))
                             

    model.add(Flatten()) 
    model.add(Dense(10, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    
    return model, model_name

    
    
def fit_eval_model (build_model, X, Y, Xt, Yt):
    '''
    
    Crea y entrena el modelo build_model con los datos de train X,Y
    y lo evalua con los datos de test Xt,Yt
    Saca la gráfica del entrenamiento 
    Devuelve los MSE de train, evaluación y test multiplicados por el factor de normalización maxdatos
    
    Si se introduce None en build_model, se predice con el valor anterior de la serie (modelo dummy)
    
    '''
    
    # Los datos de entrenamiento se desglosan en 80% entrenamiento y el 20% de evaluación
    Xtr, Xval, Ytr, Yval = train_test_split(X, Y, test_size=0.2, random_state=42)
    
    if build_model == None:
        model_name = "DUMMY"
        model = None
        acc_tr =  1 - abs(Ytr - np.roll(Ytr, shift=1)).sum()/len(Ytr)
        acc_val =  1 - abs(Yval - np.roll(Yval, shift=1)).sum()/len(Yval)
        acc_test =  1 - abs(Yt - np.roll(Yt, shift=1)).sum()/len(Yt)

        print (f"ACC modelo {model_name} en datos de train: {acc_tr}, validación: {acc_val} y test: {acc_test}")

        return model, model_name, acc_tr, acc_val, acc_test

    else: 
        if build_model != build_model_regresion and build_model != build_model_LSTM:
            model = None
            model_name = " "
            acc_tr = 0
            acc_val = 0
            acc_test = 0
            return model, model_name, acc_tr, acc_val, acc_test

        model, model_name = build_model(X)

        _learning_rate = 0.01
        _epochs = 100
        _batch_size = 32
        
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=_learning_rate), 
              loss=['mse'],
              metrics=['accuracy'])

        hist = model.fit(Xtr,
                         Ytr,
                         epochs=_epochs,
                         validation_data=(Xval, Yval),
                         batch_size=_batch_size,
                         verbose = 0)

        plt.figure(figsize=(10,6))
        plt.subplot(1,2,1)
        plt.plot(hist.history['loss'],label="loss")
        plt.plot(hist.history['val_loss'],label="val_loss")
        plt.legend()
        plt.subplot(1,2,2)
        plt.plot(hist.history['accuracy'],label="acc")
        plt.plot(hist.history['val_accuracy'],label="val_acc")
        plt.legend()

        _loss, acc_tr = model.evaluate(Xtr, Ytr, verbose=0)
        _loss, acc_val = model.evaluate(Xval, Yval, verbose=0)
        _loss, acc_test = model.evaluate(Xt, Yt, verbose=0)


        print (f"ACC modelo {model_name} en datos de train: {acc_tr}, validación: {acc_val} y test: {acc_test}")

        return model, model_name, acc_tr, acc_val, acc_test

```
