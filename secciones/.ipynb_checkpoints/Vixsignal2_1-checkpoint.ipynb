## Analizar alpha generado por la señal VIXSI en el activo SPY

Utilizando los datos históricos de la señal VIXSI, se puede analizar el alpha generado por el algoritmo de trading que utiliza la señal VIXSI para mantener comprado (VIXSI=1) o vendido (en corto) (VIXSI=-1) diariamente el activo SPY, donde el benchmark es el propio activo.

[alpha VIXSI con SPY](#alpha_vixsi_spy)



```python
#%pip install --upgrade investpy

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
```

### alpha_vixsi_spy

Una forma de validar la señal es ver cuanto alpha genera la estrategia que utiliza la señal para invertir en largo (vixsi=1) o en corto (vixsi=-1) utilizando como activo el ETF SPY, que es el replica al índice SP500.

la fórmula del alpha se basará en el alpha de Jensen (asignando 0 a la rentabilidad libre de riesgo):

rentabilidad señal - Beta_señal_activo * rentabilidad activo

El alpha se calculará por año, y se evaluará la media y el ratio de sharpe del alpha anual.

Los datos del SPY se recogerán de yfinance y se ajustarán a los dividendos.


```python
def f_alpha(datos, estrategia="estrategia", benchmark="SPY"):
    '''

    '''

    srets = np.log(datos[estrategia]).diff().fillna(0)
    brets = np.log(datos[benchmark]).diff().fillna(0)

    #w_brets = brets.loc[srets.index]
    cov_rets = np.cov(srets.T, brets)[0,1:]
    var_brets = brets.var()
    beta = cov_rets / var_brets
    rent_srets = srets.sum()
    rent_brets = brets.sum()
    alpha = rent_srets - beta * rent_brets    
    return (alpha)

benchmark = "SPY"

desde = "2014-01-01"
hasta = "2024-01-01"

df_vixsi = pd.read_csv("../datos/VIXSIyf.csv", index_col=0)
df_vixsi.index =  pd.to_datetime(df_vixsi.index)
#df_vixsi.VIXSI = (df_vixsi.VIXSI - 1)/2

spy_datos = yf.download("SPY", start=desde, end=hasta, auto_adjust=True)
datos = pd.concat([spy_datos.Close, df_vixsi.VIXSI], axis=1).dropna()
datos.columns = ["SPY","vixsi"]
datos["estrategia"] = (1 + datos.vixsi.shift(1) * datos[benchmark].pct_change()).cumprod()
años = sorted(datos.index.year.unique())
alpha = np.zeros(len(años))
for i in range(len(años)-1):
    alpha[i] = f_alpha(datos.loc[str(años[i])])
    
alpha = pd.DataFrame(alpha, index=años)

print("media anual del alpha:", alpha.mean().values)
print("Ratio de sharpe de los alpha anuales:", alpha.mean().values/alpha.std().values)
alpha.plot()
plt.axhline(y=0, color='b', linestyle='--')
```

### alphas_spy

Se crean las siguientes señales para compararlas con vixsi:

Cruce de precio con media de 9 sesiones 1 si está por encima, -1 si está por debajo
Cruce de dos medias de 9 con media de 20 sesiones 1 si está por encima, -1 si está por debajo


```python
def f_alpha_señal (datos, señal, desde, hasta):
    estrategia = "estr"+señal
    datos[estrategia] = (1 + datos[señal].shift(1) * datos["SPY"].pct_change()).cumprod()
    años = sorted(datos.index.year.unique())
    alpha = np.zeros(len(años))
    for i in range(len(años)):
        alpha[i] = f_alpha(datos.loc[str(años[i])],estrategia=estrategia)

    alpha = pd.DataFrame(alpha, index=años)
    #return (alpha, datos[estrategia][-1])
    return (alpha, alpha.mean())
    #alpha.plot()
    #plt.axhline(y=0, color='b', linestyle='--')
    
datos["signal0"] = ((datos.SPY >= datos.SPY.rolling(9).mean())*1 + (datos.SPY < datos.SPY.rolling(9).mean())*-1)
datos["signal1"] = datos["signal0"] * -1
datos["signal2"] = ((datos.SPY.rolling(20).mean() < datos.SPY.rolling(9).mean())*1 +
           (datos.SPY.rolling(20).mean() >= datos.SPY.rolling(9).mean())*-1)
datos["signal3"] = datos["signal2"]
señales =["vixsi","signal0","signal1","signal2","signal3"]

alpha_rsharpe = np.zeros(len(señales))
for i,señal in enumerate(señales):
    alpha, rsharpe = f_alpha_señal (datos, señal, desde, hasta)
    alpha_rsharpe[i] = rsharpe
    
alphas_df = pd.DataFrame(alpha_rsharpe, index=señales)
señal = alphas_df.sort_values(0, ascending=False).index[0]

print("señal con mejor alpha: ", señal)
alpha, rsharpe = f_alpha_señal (datos, señal, desde, hasta)
alpha.plot()
plt.axhline(y=0, color='b', linestyle='--')
```
