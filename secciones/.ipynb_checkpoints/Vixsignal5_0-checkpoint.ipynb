## Opciones diarias del SPY

De manera resumida, se puede decir que el Vix (Volatility Index)  mide las expectativas del mercado sobre la volatilidad implícita a 30 días en los precios ATM de las opciones del SP500 utilizando las opciones con más de 23 días y menos de 37 días hasta el vencimiento.

## Como se interpreta?

Hay varias maneras de interpretar el VIX, la más conocida es la que relaciona diferentes niveles del VIX con la situación del mercado. (hasta 20 mercado alcista, entre 20 y 30 mercado volatil y mayor que 30 mercado bajista). Sin embargo, el VIX también se puede utilizar para crear señales de sobrevaloración y infravaloración de la volatilidad del SP500.

## Creación de la señal basada en el VIX

Utilizando esta última interpretación, la de la volatilidad infravalorada del SP500, se crea una señal diaria, que es 1 por defecto excepto para los casos en los que la volatilidad está infravalorada, que será -1. Dicha señal se denominará a  partir de ahora VIXSI, y por defecto se calculará a final de día de las acciones/ETFs americanos.

## Obtención VIX en tiempo real

El valor del VIX en tiempo real se obtiene al final de dia de la cotización de las acciones/ETF americanos (1 minuto antes). Este dato se extrae por scpraping de la web "investing.com" debido a que yfinance muestra un retardo de 15 minutos.

```python
#%pip install --upgrade investpy

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
#from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time

from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense, LSTM, Flatten, BatchNormalization, MaxPooling1D, Attention
from tensorflow.keras.models import Model
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Dense, BatchNormalization
```

```python
df_op = pd.read_csv("../datos/opciones_SPY.csv", index_col=0)
#df_op.index =  pd.to_datetime(df_op.index)
```

```python
df_op.columns
```

```python
df_op["dstrike"] = np.abs(df_op.strike - df_op.active_underlying_price_1545)
df_op["fecha"] = pd.to_datetime(df_op.quote_date)
df_op["fecha_exp"] = pd.to_datetime(df_op.expiration)
df_op["hasta_exp"] = (df_op.fecha_exp - df_op.fecha).dt.total_seconds()/86400
```

```python
op_put = df_op[(df_op.option_type=="P") & (df_op.dstrike<10)]
op_call = df_op[(df_op.option_type=="C") & (df_op.dstrike<10)]
```

```python
def vol_exp (df_op, desde=None, hasta=None):
    if desde == None:
        return df_op.groupby("fecha").implied_volatility_1545.mean()
    else:
        return df_op.loc[((op_call.hasta_exp>desde) & (df_op.hasta_exp<=hasta))].groupby("fecha").implied_volatility_1545.mean()

cvolspy = vol_exp(op_call)
cvolspy0 = vol_exp(op_call, desde=0, hasta=7)
cvolspy7 = vol_exp(op_call, desde=7, hasta=14)
cvolspy14 = vol_exp(op_call, desde=14, hasta=30)
cvolspy30 = vol_exp(op_call, desde=30, hasta=60)
pvolspy = vol_exp(op_put)
pvolspy0 = vol_exp(op_put, desde=0, hasta=7)
pvolspy7 = vol_exp(op_put, desde=7, hasta=14)
pvolspy14 = vol_exp(op_put, desde=14, hasta=30)
pvolspy30 = vol_exp(op_put, desde=30, hasta=60)
```

```python
cvols = pd.concat([cvolspy0, cvolspy7, cvolspy14, cvolspy30], axis=1)
cvols.columns = ("c0","c7","c14","c30")
pvols = pd.concat([pvolspy0, pvolspy7, pvolspy14, pvolspy30], axis=1)
pvols.columns = ("p0","p7","p14","p30")
```

```python
csmile = cvols.div(cvols.sum(axis=1), axis=0)
csmile.columns = ("sc0","sc7","sc14","sc30")
psmile = pvols.div(pvols.sum(axis=1), axis=0)
psmile.columns = ("sp0","sp7","sp14","sp30")
```

```python
skew = pvolspy/(cvolspy + pvolspy)
skew0 = pvolspy0/(cvolspy0 + pvolspy0)
skew7 = pvolspy7/(cvolspy7 + pvolspy7)
skew14 = pvolspy14/(cvolspy14 + pvolspy14)
skew30 = pvolspy30/(cvolspy30 + pvolspy30)

skew = pd.concat([skew, skew0, skew7, skew14, skew30], axis=1)
skew.columns = ("skew", "skew0", "skew7", "skew14", "skew30")
```

```python
desde = cvolspy.index[0]
hasta = cvolspy.index[-1]

spy_datos = yf.download("SPY", start=desde, end=hasta)
datos = pd.concat([spy_datos.Close, cvolspy, cvolspy0, cvolspy7, cvolspy14, cvolspy30], axis=1).dropna()
datos.columns = ["SPY","ivol","ivol0","ivol7","ivol14","ivol30"]
```

```python
desde = pvolspy.index[0]
hasta = pvolspy.index[-1]

spy_datos = yf.download("SPY", start=desde, end=hasta)
#datos = pd.concat([spy_datos.Close, pvolspy, pvolspy0, pvolspy7, pvolspy14, pvolspy30], axis=1).dropna()
datos = pd.concat([spy_datos.Close, cvols, pvols, csmile, psmile, skew], axis=1).dropna()
#datos.columns = ["SPY","ivol","ivol0","ivol7","ivol14","ivol30"]
```

```python
columnas = datos.columns.to_list()
columnas[0] = "SPY"
datos.columns = columnas
```

```python
def par_vixsi (par1, ivol):
    datos["rets"] = datos.SPY.pct_change()
    datos["volret"] = datos[ivol]/np.abs(datos.rets)
    datos["vixsi"] = 0
    #datos["vixsi"].loc[(datos.rets>0) & (datos.volret<par1)] = 1
    #datos["vixsi"].loc[(datos.volret<par1)] = 1
    datos["vixsi"].loc[(datos[ivol]<par1)] = 1
    return ((1 + datos.vixsi.shift(1) * datos.rets).cumprod()[-1])

maxret = -1000
maxpar = 0
maxivol = 0
#for par in range(50,1000,10):
for par in range(0,200,1):
    for ivol in ["c0","c7","c14","c30","p0","p7","p14","p30"]:
        ret = par_vixsi(par/100, ivol)
        if ret > maxret:
            maxpar = par
            maxret = ret
            maxivol = ivol

print (maxret, maxpar, maxivol)
```

```python

def build_model_regresion(X):
    '''
    Se implementa el modelo de regresión lineal, con una sola capa densa de una neurona sin función de activación
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "REGRESION"
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(XX_tr.shape[1],XX_tr.shape[2])))
    model.add(BatchNormalization())
    model.add(keras.layers.Dense(1, activation="sigmoid"))
    
    return model, model_name

def build_model_LSTM(X):
    '''
    Se implementa el modelo LSTM con una sola capa LSTM de tamaño temporal XX.shape[1] y canales XX.shape[2]
    La salida es una capa densa de una sola neurona sin función de activación
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''

    model_name = "LSTM"
    model = keras.Sequential()

    #return_sequences=True. Devuelve como salida toda la secuencia temporal en lugar de la última para la entrada de la siguiente capa
    #Como salida devuelve la misma dimensión temporal que la entrada

    #model.add(keras.layers.LSTM(X.shape[1], input_shape=(X.shape[1],X.shape[2]), return_sequences=True))
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.LSTM(X.shape[1]*2, activation="relu", return_sequences=True))
    model.add(keras.layers.Flatten())
    model.add(Dense(X.shape[1], activation="relu"))
    model.add(Dense(1, activation="sigmoid"))

    return model, model_name

def build_model_GRU(X):
    '''
    Se implementa el modelo GRU de la misma manera que el modelo LSTM
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "GRU"
    model = keras.Sequential()

    #return_sequences=True. Devuelve como salida toda la secuencia temporal en lugar de la última para la entrada de la siguiente capa
    #Como salida devuelve la misma dimensión temporal que la entrada
    #model.add(keras.layers.GRU(X.shape[1], input_shape=(X.shape[1],X.shape[2],), return_sequences=True, activation="tanh"))
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.GRU(X.shape[1]*2, activation="relu", return_sequences=True))
    model.add(keras.layers.Flatten())
    model.add(Dense(X.shape[1], activation="relu"))
    model.add(Dense(1, activation="sigmoid"))

    return model, model_name

def build_model_CONV(X):
    '''
    Se implementa un modelo convolucional de dos capas de 16 y 8 filtros de tamaño 3 con pooling de 2 en cada capa para reducir la dimensión de las salidas
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''

    model_name = "CONV"
 
    model = keras.Sequential()
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    #model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='tanh', input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='relu', input_shape=(X.shape[1],X.shape[2])))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Convolution1D(filters=8, kernel_size=3, activation="relu"))
    model.add(keras.layers.Flatten())
    model.add(Dense(8, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
              
    return model, model_name

def build_model_CONVLSTM(X):
    '''
    Se implementa un modelo convolucional y recurrente (lstm), donde la primera capa será un modelo convolicional que como salida genere la misma
    dimensión que la entrada para que pase por un modelo convolucional 1D similar al modelo anterior
    
    Se le pasa la variable de entrada X para recoger las dimensiones temporal y activos
    
    '''
    
    model_name = "CONVLSTM"

    model = keras.Sequential()
    
    #Le añade return_sequences=True para pasar toda la secuencia temporal al la capa convolucional
    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    #model.add(keras.layers.LSTM(X.shape[1], input_shape=(X.shape[1],X.shape[2],), return_sequences=True))
    model.add(keras.layers.LSTM(X.shape[1], return_sequences=True, activation="relu"))
    model.add(keras.layers.Convolution1D(filters=16, kernel_size=3, activation='relu', input_shape=(X.shape[1],X.shape[2])))
    model.add(MaxPooling1D(2))
    model.add(keras.layers.Convolution1D(filters=8, kernel_size=3, activation="relu"))
    model.add(keras.layers.Flatten())
    model.add(Dense(8, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    
    return model, model_name

def build_model_attention (X):
    '''
    Se implementa un modelo de atención con tres capas densas (query, value y key) para transformar la salida de la capa LSTM en vectores de consulta,
    de valor y de clave como entrada en la capa attention
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "ATTENTION"

    #XX = Input((X.shape[1],X.shape[2],))
    inputs = Input(shape=(X.shape[1],X.shape[2]))
    XX = BatchNormalization()(inputs)
    out = LSTM(10, return_sequences = True, activation="relu")(XX)

    query = Dense(X.shape[1], activation="relu")(out)
    value = Dense(X.shape[1], activation="relu")(out)
    key = Dense(X.shape[1], activation="relu")(out)

    out = Attention(use_scale=False)([query,value,key])
    out = Flatten()(out)
    out = Dense(10, activation="relu")(out)
    out = Dense(1, activation="sigmoid")(out)

    model = Model(inputs=XX, outputs=out)
    
    return model, model_name

def build_model_bidireccional (X):
    '''
    Se implementa un modelo bidireccional asimétrico (GRU para forward yt LSTM para backward)
    Se le pasa la variable de entrada X para recoger las dimensiones de la entrada
    
    '''
    
    model_name = "BIDIRECCIONAL"

    model = keras.Sequential()

    forward_layer = keras.layers.GRU(20,activation='relu',return_sequences=True)
    backward_layer = keras.layers.LSTM(20, activation='relu',return_sequences=True, go_backwards=True)

    model.add(BatchNormalization(input_shape=(X.shape[1],X.shape[2])))
    model.add(keras.layers.Bidirectional(forward_layer, backward_layer=backward_layer,
                                         #input_shape=(X.shape[1],X.shape[2])
             ))
                             

    model.add(Flatten()) 
    model.add(Dense(10, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    
    return model, model_name

    
    
def fit_eval_model (build_model, X, Y, Xt, Yt):
    '''
    
    Crea y entrena el modelo build_model con los datos de train X,Y
    y lo evalua con los datos de test Xt,Yt
    Saca la gráfica del entrenamiento 
    Devuelve los MSE de train, evaluación y test multiplicados por el factor de normalización maxdatos
    
    Si se introduce None en build_model, se predice con el valor anterior de la serie (modelo dummy)
    
    '''
    
    # Los datos de entrenamiento se desglosan en 80% entrenamiento y el 20% de evaluación
    Xtr, Xval, Ytr, Yval = train_test_split(X, Y, test_size=0.2, random_state=42)
    
    if build_model == None:
        model_name = "DUMMY"
        model = None
        acc_tr =  1 - abs(Ytr - np.roll(Ytr, shift=1)).sum()/len(Ytr)
        acc_val =  1 - abs(Yval - np.roll(Yval, shift=1)).sum()/len(Yval)
        acc_test =  1 - abs(Yt - np.roll(Yt, shift=1)).sum()/len(Yt)

        print (f"ACC modelo {model_name} en datos de train: {acc_tr}, validación: {acc_val} y test: {acc_test}")

        return model, model_name, acc_tr, acc_val, acc_test

    else: 
        if build_model != build_model_regresion and build_model != build_model_LSTM:
            model = None
            model_name = " "
            acc_tr = 0
            acc_val = 0
            acc_test = 0
            return model, model_name, acc_tr, acc_val, acc_test

        model, model_name = build_model(X)

        _learning_rate = 0.01
        _epochs = 100
        _batch_size = 32
        
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=_learning_rate), 
              loss=['mse'],
              metrics=['accuracy'])

        hist = model.fit(Xtr,
                         Ytr,
                         epochs=_epochs,
                         validation_data=(Xval, Yval),
                         batch_size=_batch_size,
                         verbose = 0)

        plt.figure(figsize=(10,6))
        plt.subplot(1,2,1)
        plt.plot(hist.history['loss'],label="loss")
        plt.plot(hist.history['val_loss'],label="val_loss")
        plt.legend()
        plt.subplot(1,2,2)
        plt.plot(hist.history['accuracy'],label="acc")
        plt.plot(hist.history['val_accuracy'],label="val_acc")
        plt.legend()

        _loss, acc_tr = model.evaluate(Xtr, Ytr, verbose=0)
        _loss, acc_val = model.evaluate(Xval, Yval, verbose=0)
        _loss, acc_test = model.evaluate(Xt, Yt, verbose=0)


        print (f"ACC modelo {model_name} en datos de train: {acc_tr}, validación: {acc_val} y test: {acc_test}")

        return model, model_name, acc_tr, acc_val, acc_test

```

```python
# Se crea un dataframe donde se guardarán los resultados de todos los modelos
models = []
nombre_modelos = np.arange(8).tolist()
result = pd.DataFrame(np.zeros((8,3)))
result.columns = ["acc_train","acc_val","acc_test"]
```
