```python
# Se optimizan los precios con un delta operador para conseguir unos drawdowns objetivo
# y se generan los trades con dichos precios óptimos con el siguiente valor no visto siguiendo 
# con un walk forward de ventana expansiva con un único dato de test (el valor siguiente)

# El problema de esta versión es que la optimización es direccional, y depende de si el periodo es alcista o bajista
# La parte positiva es que las compras tienen menos volatilidad, con lo que es probable que cuando el mercado baje, 
# invertir en compras mejore al mercado



import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from datetime import timedelta
from dateutil.relativedelta import relativedelta
import time
import pandas as pd
import numpy as np
import random
import time
from pathlib import Path

def get_delta (datos, f_vixsi, v_vixsi):
    '''

    '''

    qvixsi_df = pd.DataFrame([datos[f_vixsi].rank(pct=True)])
    qvixsi_df.columns = ["QVIXSI"]
    if v_vixsi >= datos[f_vixsi].min():
            qvixsi = qvixsi_df.loc[datos[f_vixsi] < v_vixsi].QVIXSI.max()
    else:
            qvixsi = 0
        
    return qvixsi

def ini_deltavix (desde, hasta):
    '''
    
    '''

    spy_datos = yf.download("SPY", start=desde, end=hasta)
    vix_datos = yf.download("^VIX", start=desde, end=hasta)

    datos_all = pd.concat([spy_datos.Close, vix_datos.Close], axis=1).dropna()
    datos_all.columns = ["SPY","VIX"]
    datos_all["rets"] = datos_all.SPY.pct_change()

    # Indicadores para deltavix
    datos_all["FVIX"] = datos_all.VIX
    datos_all["FRETVIX"] = datos_all.VIX.pct_change()
    datos_all["FRET"] = datos_all.rets
    datos_all["FRVIX"] = np.abs(datos_all.rets) - datos_all.VIX/(np.sqrt(256)*100)

    datos_all.dropna(inplace=True)

    return datos_all

def get_deltavix (datos_all, desde, hasta, qrank=False):
    
    datos = datos_all.iloc[desde_train:hasta_train].copy()    
    if qrank:
        datos["qretvix"] = datos.FRETVIX.rank(pct=True)
        datos["qret"] = datos.FRET.rank(pct=True)
        datos["qrvix"] = datos.FRVIX.rank(pct=True)
        datos["qvix"] = datos.FVIX.rank(pct=True)
    else:
        #datos["qretvix"] = (datos.FRETVIX - datos.FRETVIX.min())/(datos.FRETVIX.max() - datos.FRETVIX.min())
        #datos["qret"] = (datos.FRET - datos.FRET.min())/(datos.FRET.max() - datos.FRET.min())
        #datos["qrvix"] = (datos.FRVIX - datos.FRVIX.min())/(datos.FRVIX.max() - datos.FRVIX.min())
        #datos["qvix"] = (datos.FVIX - datos.FVIX.min())/(datos.FVIX.max() - datos.FVIX.min())
        
        datos["qretvix"] = datos.FRETVIX
        datos["qret"] = datos.FRET
        datos["qrvix"] = datos.FRVIX
        datos["qvix"] = datos.FVIX

    return datos

def get_options ():
    
    df_op202212 = pd.read_csv("opciones/SPY2022-12.csv", index_col=0, parse_dates=True)

    df_op202301 = pd.read_csv("opciones/SPY2023-01.csv", index_col=0, parse_dates=True)
    df_op202302 = pd.read_csv("opciones/SPY2023-02.csv", index_col=0, parse_dates=True)
    df_op202303 = pd.read_csv("opciones/SPY2023-03.csv", index_col=0, parse_dates=True)
    df_op202304 = pd.read_csv("opciones/SPY2023-04.csv", index_col=0, parse_dates=True)
    df_op202305 = pd.read_csv("opciones/SPY2023-05.csv", index_col=0, parse_dates=True)
    df_op202306 = pd.read_csv("opciones/SPY2023-06.csv", index_col=0, parse_dates=True)
    df_op202307 = pd.read_csv("opciones/SPY2023-07.csv", index_col=0, parse_dates=True)
    df_op202308 = pd.read_csv("opciones/SPY2023-08.csv", index_col=0, parse_dates=True)
    df_op202309 = pd.read_csv("opciones/SPY2023-09.csv", index_col=0, parse_dates=True)
    df_op202310 = pd.read_csv("opciones/SPY2023-10.csv", index_col=0, parse_dates=True)
    df_op202311 = pd.read_csv("opciones/SPY2023-11.csv", index_col=0, parse_dates=True)
    df_op202312 = pd.read_csv("opciones/SPY2023-12.csv", index_col=0, parse_dates=True)

    df_op202401 = pd.read_csv("opciones/SPY2024-01.csv", index_col=0, parse_dates=True)
    df_op202402 = pd.read_csv("opciones/SPY2024-02.csv", index_col=0, parse_dates=True)
    df_op202403 = pd.read_csv("opciones/SPY2024-03.csv", index_col=0, parse_dates=True)
    df_op202404 = pd.read_csv("opciones/SPY2024-04.csv", index_col=0, parse_dates=True)
    df_op202405 = pd.read_csv("opciones/SPY2024-05.csv", index_col=0, parse_dates=True)
    df_op202406 = pd.read_csv("opciones/SPY2024-06.csv", index_col=0, parse_dates=True)
    df_op202407 = pd.read_csv("opciones/SPY2024-07.csv", index_col=0, parse_dates=True)
    df_op202408 = pd.read_csv("opciones/SPY2024-08.csv", index_col=0, parse_dates=True)
    df_op202409 = pd.read_csv("opciones/SPY2024-09.csv", index_col=0, parse_dates=True)
    df_op202410 = pd.read_csv("opciones/SPY2024-10.csv", index_col=0, parse_dates=True)
    df_op202411 = pd.read_csv("opciones/SPY2024-11.csv", index_col=0, parse_dates=True)
    df_op = pd.concat([          
                      df_op202212,                   
                      df_op202301,
                      df_op202302,
                      df_op202303,
                      df_op202304,
                      df_op202305,
                      df_op202306,
                      df_op202307,
                      df_op202308,
                      df_op202309,
                      df_op202310,
                      df_op202311,                   
                      df_op202312,                   
                      df_op202401,
                      df_op202402,
                      df_op202403,
                      df_op202404,
                      df_op202405,
                      df_op202406,
                      df_op202407,
                      df_op202408,
                      df_op202409,
                      df_op202410,
                      df_op202411                   
                      ], axis=0)
    
    df_op["fecha"] = pd.to_datetime(df_op["quote_date"])
    df_op["fexp"] = pd.to_datetime(df_op["expiration"])
    df_op["indice"] = df_op["fexp"]
    df_op = df_op[(df_op["quote_date"]!=df_op["expiration"])]
    df_op['delta'] = (df_op['fexp'] - df_op['fecha']).dt.days
    df_op.reset_index(inplace=True, drop=True)
    df_op.set_index('indice', inplace=True)
    df_op = df_op.loc[df_op.groupby('fecha')['delta'].idxmin()]

    df_op["CATM"] = df_op['underlying_ask_eod'].astype(int)
    df_op["resto"] = df_op['underlying_bid_eod']%1 > 0
    df_op["PATM"] = df_op['underlying_bid_eod'].astype(int) + df_op["resto"] * 1

    df_op["CSTRIKE"] = df_op.strike - df_op.CATM
    df_op["PSTRIKE"] = df_op.strike - df_op.PATM
    df_op = df_op[np.abs(df_op.CSTRIKE)<=50]
    df_op = df_op[["fecha","fexp","underlying_ask_eod",
                        "option_type","strike","bid_eod","ask_eod","CSTRIKE","PSTRIKE"]]
    df_op.columns=["FECHA","FEXP","ACTIVO","TIPO","STRIKE","BID","ASK","CSTRIKE","PSTRIKE"]
    
    df_opc = df_op.loc[(df_op.CSTRIKE == 0) & (df_op.TIPO=="C")].copy()
    df_opc['delta'] = (df_opc['FEXP'] - df_opc['FECHA']).dt.days
    df_opc.reset_index(inplace=True, drop=True)
    df_opc = df_opc.loc[df_opc.groupby('FECHA')['delta'].idxmin()]
    df_opc.set_index('FEXP', inplace=True)
    df_opc["cpuntos"] = df_opc["ACTIVO"].shift(-1) - df_opc["STRIKE"]
    df_opc["ccgain"] = np.where(df_opc.cpuntos > 0, df_opc.cpuntos - df_opc.ASK, - df_opc.ASK) 
    df_opc["vcgain"] = df_opc.ASK
    df_opc["vcgain"] = np.where(df_opc.cpuntos > 0, df_opc.ASK - df_opc.cpuntos, df_opc.ASK)   
    df_opc = df_opc[["ccgain", "vcgain", "cpuntos"]]
    
    df_opp = df_op.loc[(df_op.PSTRIKE == 0) & (df_op.TIPO=="P")].copy()
    df_opp['delta'] = (df_opp['FEXP'] - df_opp['FECHA']).dt.days
    df_opp.reset_index(inplace=True, drop=True)
    df_opp = df_opp.loc[df_opp.groupby('FECHA')['delta'].idxmin()]
    df_opp.set_index('FEXP', inplace=True)
    df_opp["ppuntos"] = df_opp["ACTIVO"].shift(-1) - df_opp["STRIKE"]
    df_opp["cpgain"] = np.where(df_opp.ppuntos < 0, - df_opp.ppuntos - df_opp.BID, - df_opp.BID) 
    df_opp["vpgain"] = df_opp.BID
    df_opp["vpgain"] = np.where(df_opp.ppuntos < 0, df_opp.BID + df_opp.ppuntos, df_opp.BID)   
    df_opp = df_opp[["cpgain", "vpgain", "ppuntos"]]
    
    return df_opc, df_opp
```

```python
desde = "2008-01-01"
hasta = (datetime.now() + timedelta(1)).strftime("%Y-%m-%d")
hasta = "2024-11-20"
datos = ini_deltavix (desde, hasta)

#num_dias_test = len(datos.loc[df_opc.index[0]:])
num_dias_test = len(datos.loc["2022-05-01":])

ini = time.time()
df_qvixsi = pd.DataFrame()
for num_dias in range(num_dias_test):
    hasta_train = len(datos) - (num_dias_test - num_dias + 1)
    desde_train = 0
    
    datos_train = get_deltavix (datos, desde_train, hasta_train)
    datos["qretvix"] = None
    datos["qret"] = None
    datos["qrvix"] = None
    datos["qvix"] = None
    
    fecha = datos_train.index[-1]
    fechasig = datos.index[hasta_train]
    df_qvixsi1 = pd.DataFrame([[fecha, fechasig, 
                               datos_train.iloc[-1].qretvix,
                               datos_train.iloc[-1].qret,
                               datos_train.iloc[-1].qrvix,
                               datos_train.iloc[-1].qvix]])
    df_qvixsi1.columns = ["fecha", "fechasig", "qretvix", "qret", "qrvix", "qvix"]
    df_qvixsi1.set_index("fechasig", inplace=True)
    df_qvixsi = pd.concat([df_qvixsi, df_qvixsi1], axis=0)
```

```python

def optimizar (trades, opgain):
    
    qtrades = trades.copy()
    qtrades["vpoutl"] = 0
    qtrades.loc[qtrades[opgain] < qtrades[opgain].quantile(0.05),"vpoutl"] = 1
    minmax = qtrades.loc[qtrades.vpoutl==1].qvixsi.describe()
    optmin = minmax[3]
    optmax = minmax[7]

    return optmax, optmin

```

```python
# Optimización utilizando datos de train (el 80%) y datos de test (el 20% restante)

def optimizar_train_test (datos, df_result, opgain, qvixsi, ini_test, fin_test, in_test_train=True):
    '''
    
    '''

    if in_test_train:
        datos_train = datos.iloc[0:fin_test].copy()
    else:
        datos_train = datos.iloc[0:ini_test].copy()
        
    datos_test = datos.iloc[ini_test:fin_test].copy()

    if qvixsi != "None":      
        datos_train["qvixsi"] = datos_train[qvixsi]
        datos_test["qvixsi"] = datos_test[qvixsi]

        optmax, optmin = optimizar (datos_train, opgain=opgain)

        #if len(df_result) > 0:
        #    df_result = df_result[df_result.SIGNAL==qvixsi]
        #    num_regs = len(df_result)
        #    if num_regs > 0:
        #        optmax = (df_result.MAX.sum() + optmax)/(num_regs + 1)
        #        optmin = (df_result.MIN.sum() + optmin)/(num_regs + 1)

        qvtrades = datos_test
        qvtrades.loc[(qvtrades.qvixsi <= optmax) &
                     (qvtrades.qvixsi >= optmin), opgain] = 0
    else:
        qvtrades = datos_test
        optmax, optmin = 1, 0 

    df_result = pd.DataFrame()
    df_result = pd.DataFrame([[qvixsi, opgain, ini_test, fin_test, optmax, optmin]])
    df_result.columns = ["SIGNAL", "OPGAIN", "INI_TEST", "FIN_TEST", "MAX","MIN"]

    return df_result, qvtrades
```

```python
#Para un test del 10% en cualquier lugar del periodo hacia adelante desde el 40% de los datos

def optimizar_wfe (trades, opgain, min_dias_train=250, dias_test=20):
    
    df_result = pd.DataFrame()
    df_None = pd.DataFrame()
    df_qrvix = pd.DataFrame()
    df_qvix = pd.DataFrame()
    df_qret = pd.DataFrame()
    df_qretvix = pd.DataFrame()

    maxregs = len (trades)
    ini_test = min_dias_train
    while (ini_test < maxregs):
        fin_test = ini_test + dias_test
        if fin_test >= maxregs:
            fin_test = maxregs


        result, strades = optimizar_train_test (trades, df_result, opgain, qvixsi="qvix",
                                                ini_test=ini_test, fin_test=fin_test)
        df_result  = pd.concat([df_result, result])
        df_qvix = pd.concat([df_qvix,strades])

        result, strades = optimizar_train_test (trades, df_result, opgain, qvixsi="qrvix",
                                                ini_test=ini_test, fin_test=fin_test)
        df_result  = pd.concat([df_result, result])
        df_qrvix = pd.concat([df_qrvix,strades])
        
        ini_test = fin_test

    return df_result, {"vix":df_qvix, "rvix":df_qrvix}
```

```python
ini = time.time()
res_vc, eq_vc = optimizar_wfe (trades, "vcgain", min_dias_train=250, dias_test=1)
res_vp, eq_vp = optimizar_wfe (trades, "vpgain", min_dias_train=250, dias_test=1)
print ("tiempo proceso:", time.time() - ini)
```
