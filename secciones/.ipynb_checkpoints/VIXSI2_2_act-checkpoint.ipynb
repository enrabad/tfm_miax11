### Creación cartera para actuar en corto en la estrategia VIXSI

Se ha visto en anteriores notebooks que el alpha generado por la extrategia VIXSI recae en la señal Vixsi -1, y que en la versión ponerse en corto, en promedio, cerca de la mitad de los activos del SP500 contienen un alfa superior al SPY, con lo que aparentemente no parece difícil encontrar una cartera de acciones que contenga un alfa superior al SPY (a partir de ahora, cuando se hable de alpha de un activo, siempre se referirá al alpha generado por la estrategia VIXSI en ese activo, que equivaldrá al doble de la rentabilidad de la señal en corto).

En este notebook se expone primero un ejemplo de cobertura ideal ideal, el de una cartera formada únicamente por el activo con mejor ratio de sharpe de alphas, y se compara con la estrategia VIXSI. 

Luego se intenta encontrar una cartera de K activos, utilizando como criterio de selección el diferencial de alpha entre SPY y el activo (a partir de ahora se denominará dalpha), seleccionando los K activos con mejores dalphas.

El dataset de dalphas se dividirá en dos periodos, uno para seleccionar los activos, denominado periodo train, y otro para probarlos con las estrategia, denominado periodo test. 

Al final se hará un resumen de los resultados obtenidos


```python
#Importamos las librerias necesarias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import yfinance as yf
import requests

%matplotlib inline
%load_ext autoreload
%autoreload 2
import warnings
warnings.filterwarnings('ignore')

from modulos.tfm_util import metrica_cartera
from modulos.tfm_util import drwdown
```

### Cargar señal vixi y alphas anualizados

```python
datos_vixsi = pd.read_csv ("../datos/vixsi.csv", index_col = 0)
datos_vixsi.index = pd.to_datetime(datos_vixsi.index)

df_alpha = pd.read_csv("../datos/alphas_ann.csv", index_col=0)
```

### Se selecciona el activo para la cartera de cobertura ideal

```python
activo = (df_alpha.mean()/df_alpha.std()).idxmax()
```

### Se cargan los datos activos y SPY para crear la estrategia VIXSI mixta

```python
datos = yf.download([activo,"SPY"], datos_vixsi.index[0], datos_vixsi.index[-1] + pd.Timedelta(days=1))['Close']
```

### Se crea la equity de la estrategia VIXSI mixta con KLAC

```python
datos["vixsignal"] = datos_vixsi.VIXSI
datos["VIXSIhedge"] = (1 + ((datos.vixsignal.shift(1) + 1)/2) * datos[activo].pct_change() +
                        ((datos.vixsignal.shift(1) - 1)/2) * datos[activo].pct_change()).cumprod()
datos["VIXSI"] = (1 + datos.vixsignal.shift(1) * datos.SPY.pct_change()).cumprod()
```

### Se analizan los resultados

La estrategia vixsihedge con el activo elegido mejora en todos los indicadores a la estrategia VIXSI. Sin embargo, se ha elegido este activo a posterori. El objetivo será encontrar estos activos con algún método de selección de activos a priori.

```python
estrategias = ["VIXSIhedge","VIXSI"]

ret_ann, ret_ann_b, r_sharpe, r_sharpe_b = metrica_cartera(datos[estrategias[0]], datos[estrategias[1]])
maxdwn, *_ = drwdown (datos[estrategias[0]])
maxdwn_b, *_ = drwdown (datos[estrategias[1]])

valores = [[ret_ann, r_sharpe, maxdwn], [ret_ann_b, r_sharpe_b, maxdwn_b]]
df_result = pd.DataFrame(valores, index= estrategias)
df_result.columns = ["RetAnual","RSharpe","MaxDwd"]

plt.figure(figsize=(10,6))
plt.plot(datos[estrategias[0]], label=estrategias[0])
plt.plot(datos[estrategias[1]], label=estrategias[1])

plt.legend()
plt.plot()
#plt.savefig('estrategia1.png')

df_result = df_result.round(2)
df_result.to_clipboard()
df_result
```

### Cargar datos y alphas relativos a SPY

Se cargan los alphas restandole el alpha de SPY para obtener sus valores relativos, ya que lo que interesa son activos que mejoren el alpha de SPY.

```python

data = pd.read_csv("../datos/sp500.csv", index_col=0)
data.index = pd.to_datetime(data.index)

indices = list(set(data.index) & set(datos_vixsi.index))
data = data.loc[indices]
data.sort_index(inplace=True)

#solo interesan los retornos
data = data.pct_change()

df_alpha = pd.read_csv("../datos/alphas_ann.csv", index_col=0)
#valor relativo del alpha
df_alpha = df_alpha.sub(df_alpha.SPY, axis=0)

df_alpha = df_alpha[list(set(data.columns) - {"SPY"})]
df_alpha = pd.read_csv("../datos/alphas_ann.csv", index_col=0)

df_alpha = df_alpha.sub(df_alpha.SPY, axis=0)
df_alpha = df_alpha[list(set(data.columns) - {"SPY"})]

# Se incluye la señal desplazada, cargada al inicio del notebook
data["vixsignal"] = datos_vixsi.VIXSI.shift(1)
```

### Selección de activos en periodo train

Se va a selecciar los mejores activos en término de alpha relativa en el periodo de train 2008:2018, donde se seleccionarán los mejores K activos (metaparámetro predeterminado), y el criterio de selección de mejor ratio de sharpe

```python
def mejor_dalpha_rsh (dalphas, k):
    '''
    Selecciona los k activos con el ratio media/var más altos
    
    '''
    
    return list((dalphas.mean()/dalphas.std()).sort_values(ascending=False).index[:k])

desde_train = "2008"
hasta_train = "2018"

# función de selección
fsel = mejor_dalpha_rsh

# número de activos
k = 3

tr_alpha = df_alpha.loc[desde_train:hasta_train]
activos_opt = fsel(tr_alpha, k)
```

### Probar estrategia con la cartera generada en el periodo train

Comparación del alpha de la cartera combinada con el alpha del SPY periodo 2019 en adelante


```python
desde_test = "2019"
hasta_test = "2024"

retornos = np.array (data.loc[desde_test:hasta_test, activos_opt])
señal = np.array (data.loc[desde_test:hasta_test].vixsignal).reshape(-1,1)
señal_corto = np.array ((data.loc[desde_test:hasta_test].vixsignal-1)/2).reshape(-1,1)
ret_hedge_corto = (retornos * señal_corto * 1/len(activos_opt)).sum(axis=1)
ret_hedge_largo = np.array((data.loc[desde_test:hasta_test].vixsignal+1)/2 * data.loc[desde_test:hasta_test].SPY)
ret_vixsihedge = ret_hedge_corto + ret_hedge_largo
ret_vixsi = data.SPY * data.vixsignal
```

```python
tdatos = data.loc[desde_test:hasta_test]
tdatos["VIXSIhedge"] = (1 + ret_vixsihedge).cumprod()
tdatos["VIXSI"] = (1 + ret_vixsi).cumprod()

estrategias = ["VIXSIhedge", "VIXSI"]

tdatos = tdatos.loc["2019":,estrategias]
tdatos[estrategias] = tdatos[estrategias]/tdatos[estrategias].iloc[0]
#tdatos["VIXSIMIXCART"] = tdatos.VIXSIMIXCART/tdatos.VIXSIMIXCART.iloc[0]
#tdatos["VIXSISPY"] = tdatos.VIXSISPY/tdatos.VIXSISPY.iloc[0]

ret_ann, ret_ann_b, r_sharpe, r_sharpe_b = metrica_cartera(tdatos[estrategias[0]], tdatos[estrategias[1]])
maxdwn, *_ = drwdown (tdatos[estrategias[0]])
maxdwn_b, *_ = drwdown (tdatos[estrategias[1]])

valores = [[ret_ann, r_sharpe, maxdwn], [ret_ann_b, r_sharpe_b, maxdwn_b]]
df_result = pd.DataFrame(valores, index= estrategias)
df_result.columns = ["RetAnual","RSharpe","MaxDwd"]

plt.figure(figsize=(10,6))
plt.plot(tdatos[estrategias[0]], label=estrategias[0])
plt.plot(tdatos[estrategias[1]], label=estrategias[1])

plt.legend()
plt.plot()
#plt.savefig('estrategia1.png')

df_result = df_result.round(2)
df_result.to_clipboard()
df_result
```

### Conclusiones

Con una selección de activos permanente se puede crear una estrategia vixsi mixta que mejore la rentabilidad de la estrategia vixsi del SPY, aunque en este caso la mejora no ha sido relevante.

En el periodo test, en lugar de obtener los activos, se obtendrán los metaparámetros para evaluarlos en un nuevo periodo de train que incluya también el periodo de evaluación, y así cada año hasta finalizar el histórico.
